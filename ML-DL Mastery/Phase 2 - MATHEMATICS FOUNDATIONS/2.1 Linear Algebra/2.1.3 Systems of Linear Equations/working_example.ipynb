# 2.1.3 Systems of Linear Equations

Solving Ax=b: np.linalg.solve, lstsq, Gaussian elimination, condition numbers.

```python
import numpy as np
# Square system
A=np.array([[2.,1.,-1.],[-3.,-1.,2.],[-2.,1.,2.]])
b=np.array([8.,-11.,-3.])
x=np.linalg.solve(A,b)
print('x =',x)
print('Ax-b =',A@x-b)
```

```python
# Overdetermined (least squares) - Cal Housing OLS
import csv, urllib.request; from pathlib import Path
DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'cal_housing.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)
with open(dest) as f: rows=list(csv.DictReader(f))
X=np.array([[float(r[c]) for c in ['MedInc','HouseAge','AveRooms']] for r in rows[:150]])
y=np.array([float(r['MedHouseVal']) for r in rows[:150]])
Xb=np.hstack([np.ones((len(X),1)),X])
w,_,_,sv=np.linalg.lstsq(Xb,y,rcond=None)
yh=Xb@w; print(f'R² = {1-np.sum((y-yh)**2)/np.sum((y-y.mean())**2):.4f}')
```

```python
# Condition number
for s in [1,1e6,1e12]:
    A=np.array([[1.,s],[1.,s+1.]])
    print(f'scale={s:.0e}  cond={np.linalg.cond(A):.2e}')
```

```python
# Underdetermined (infinite solutions) - use lstsq for min-norm solution
A=np.array([[1.,2.,3.],[4.,5.,6.]])
b=np.array([14.,32.])
w,_,_,_=np.linalg.lstsq(A,b,rcond=None)
print('Min-norm solution:',w.round(4))
print('Check Aw=b:',np.allclose(A@w,b))
```
