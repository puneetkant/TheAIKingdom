# 2.1.6 Eigenvalues and Eigenvectors

Eigendecomposition: PCA on Cal Housing, power iteration, Markov chain stationary distribution.

```python
import numpy as np, matplotlib.pyplot as plt
A=np.array([[4.,2.],[1.,3.]])
vals,vecs=np.linalg.eig(A)
print('Eigenvalues:',vals)
print('Eigenvectors:\n',vecs)
for i in range(2): print(f'Av-λv check [{i}]:',( A@vecs[:,i]-vals[i]*vecs[:,i]).round(10))
```

```python
# PCA on Cal Housing
import csv, urllib.request; from pathlib import Path
DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'cal_housing.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)
with open(dest) as f: rows=list(csv.DictReader(f))
feat=['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup']
X=np.array([[float(r[c]) for c in feat] for r in rows[:300]])
Xc=X-X.mean(0); cov=Xc.T@Xc/(len(X)-1)
vals,vecs=np.linalg.eigh(cov)
idx=np.argsort(vals)[::-1]; vals,vecs=vals[idx],vecs[:,idx]
print('Explained var:',( vals/vals.sum()).round(4))
```

```python
# PCA scatter
W=vecs[:,:2]; Xp=Xc@W
fig,ax=plt.subplots(figsize=(6,4))
ax.scatter(Xp[:,0],Xp[:,1],alpha=0.3,s=6,c='steelblue')
ax.set_xlabel('PC1');ax.set_ylabel('PC2');ax.set_title('PCA Cal Housing');plt.show()
```

```python
# Markov chain stationary distribution
P=np.array([[0.7,0.2,0.1],[0.3,0.5,0.2],[0.1,0.3,0.6]])
vals2,vecs2=np.linalg.eig(P.T)
idx=np.argmin(np.abs(vals2-1)); pi=vecs2[:,idx].real; pi/=pi.sum()
print('Stationary π =',pi.round(4))
print('P^100 row 0 =',np.linalg.matrix_power(P,100)[0].round(4))
```
