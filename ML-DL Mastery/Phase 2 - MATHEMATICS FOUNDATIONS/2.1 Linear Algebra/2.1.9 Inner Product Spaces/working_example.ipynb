# 2.1.9 Inner Product Spaces

Inner products, Cauchy-Schwarz, kernel trick (linear/poly/RBF), Fourier projection.

```python
import numpy as np
x=np.array([1.,2.,3.]); y=np.array([4.,5.,6.])
print('<x,y>=',np.dot(x,y))
cos=np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
print('Angle:',round(np.degrees(np.arccos(cos)),2),'°')
print('Cauchy-Schwarz:',abs(np.dot(x,y)),'<=',np.linalg.norm(x)*np.linalg.norm(y))
```

```python
# Kernel functions
def k_lin(a,b): return np.dot(a,b)
def k_poly(a,b,c=1,d=2): return (np.dot(a,b)+c)**d
def k_rbf(a,b,g=0.5): return np.exp(-g*np.dot(a-b,a-b))
x=np.array([1.,2.]); y=np.array([3.,1.])
print('linear:',k_lin(x,y),'poly:',k_poly(x,y),'rbf:',round(k_rbf(x,y),6))
```

```python
# Gram matrix (RBF)
import matplotlib.pyplot as plt
X=np.random.randn(8,2)
K=np.array([[k_rbf(X[i],X[j]) for j in range(8)] for i in range(8)])
fig,ax=plt.subplots(figsize=(5,4))
im=ax.imshow(K,cmap='Blues'); plt.colorbar(im); ax.set_title('RBF Gram Matrix'); plt.show()
print('PSD:', all(np.linalg.eigvalsh(K)>=-1e-10))
```

```python
# Projection onto Fourier basis
t=np.linspace(0,2*np.pi,100)
f=np.sin(t)+0.5*np.cos(2*t)+0.1*np.random.randn(100)
B=np.column_stack([np.ones_like(t),np.sin(t),np.cos(t),np.sin(2*t),np.cos(2*t)])
c=np.linalg.solve(B.T@B,B.T@f)
fig,ax=plt.subplots(figsize=(8,3))
ax.plot(t,f,alpha=0.4,label='signal'); ax.plot(t,B@c,lw=2,label='projection')
ax.legend(); plt.show(); print('coeffs:',c.round(3))
```
