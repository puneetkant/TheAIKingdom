# 2.1.8 Matrix Decompositions (Additional)

QR, Cholesky, LU decompositions with ML applications.

```python
import numpy as np
from numpy.linalg import qr, cholesky, solve
A=np.array([[1.,2.],[4.,5.],[7.,8.]])
Q,R=qr(A); print('Q:',Q.shape,'R:',R.shape)
print('Q orthonormal:',(Q.T@Q).round(6))
print('Reconstruction err:',np.linalg.norm(A-Q@R).round(10))
```

```python
# QR least squares
b=np.array([14.,32.,50.])
x=solve(R,Q.T@b)
print('QR LS x:',x.round(4))
print('numpy lstsq:', np.linalg.lstsq(A,b,rcond=None)[0].round(4))
```

```python
# Cholesky: sample from multivariate normal
X=np.random.randn(100,3)
cov=X.T@X/100 + 1e-6*np.eye(3)
L=cholesky(cov)
print('L:\n',L.round(4))
print('LL^T=Cov:',np.allclose(L@L.T,cov))
samples=(np.random.randn(5,3)@L.T)
print('5 samples:\n',samples.round(4))
```

```python
# LU via scipy (if available)
try:
    from scipy.linalg import lu
    A=np.array([[2.,1.,-1.],[-3.,-1.,2.],[-2.,1.,2.]])
    P,L,U=lu(A)
    print('LU decomp — err:',np.linalg.norm(P@L@U-A))
except ImportError: print('scipy not available')
```
