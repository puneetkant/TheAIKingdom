# 2.1.7 Singular Value Decomposition (SVD)

SVD for PCA, low-rank approximation, pseudo-inverse, condition number.

```python
import numpy as np, matplotlib.pyplot as plt, csv, urllib.request
from pathlib import Path
DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'cal_housing.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)
with open(dest) as f: rows=list(csv.DictReader(f))
feat=['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup']
X=np.array([[float(r[c]) for c in feat] for r in rows[:300]])
X=(X-X.mean(0))/(X.std(0)+1e-9)
U,s,Vt=np.linalg.svd(X,full_matrices=False)
print(f'U:{U.shape} s:{s.shape} Vt:{Vt.shape}')
print('Singular values:',s.round(3))
```

```python
# Low-rank approximation
total=np.sum(s**2)
for k in [1,2,3,6]:
    Xk=U[:,:k]@np.diag(s[:k])@Vt[:k,:]
    print(f'k={k}: err={np.linalg.norm(X-Xk,"fro"):.4f} var={np.sum(s[:k]**2)/total:.4f}')
```

```python
# PCA via SVD scatter
scores=U*s
fig,ax=plt.subplots(figsize=(6,4))
ax.scatter(scores[:,0],scores[:,1],alpha=0.3,s=6,c='coral')
ax.set_xlabel('PC1');ax.set_ylabel('PC2');ax.set_title('SVD PCA');plt.show()
```

```python
# Pseudo-inverse
A=np.random.randn(10,4)
b=np.random.randn(10)
w=np.linalg.pinv(A)@b
print('Pseudo-inv w:',w.round(4))
# Condition number
_,s2,_=np.linalg.svd(A)
print('Condition number:',s2[0]/s2[-1])
```
