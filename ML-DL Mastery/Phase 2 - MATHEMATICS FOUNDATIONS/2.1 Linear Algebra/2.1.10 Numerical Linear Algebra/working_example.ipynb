# 2.1.10 Numerical Linear Algebra

Condition numbers, QR vs normal equations stability, sparse matrices, Jacobi iteration.

```python
import numpy as np
# Condition number comparison
A_good=np.random.randn(50,10)
A_bad=np.array([[1./(i+j+1) for j in range(10)] for i in range(50)])
for name,A in [('well',A_good),('ill',A_bad)]:
    _,s,_=np.linalg.svd(A,full_matrices=False)
    print(f'{name}: cond={s[0]/s[-1]:.2e}')
```

```python
# Normal eqn vs QR accuracy on ill-conditioned A
A=np.array([[1./(i+j+1) for j in range(10)] for i in range(50)])
b=A@np.ones(10)
x_ne=np.linalg.solve(A.T@A,A.T@b)
Q,R=np.linalg.qr(A); x_qr=np.linalg.solve(R,Q.T@b)
print('NE err:',np.linalg.norm(x_ne-np.ones(10)))
print('QR err:',np.linalg.norm(x_qr-np.ones(10)))
```

```python
# Jacobi iteration
A=np.array([[10.,-1.,2.],[-1.,11.,-1.],[2.,-1.,10.]])
b=np.array([6.,25.,-11.]); x=np.zeros(3)
D,R=np.diag(np.diag(A)),A-np.diag(np.diag(A))
for _ in range(50): x=np.linalg.solve(np.diag(np.diag(A)),b-R@x)
print('Jacobi solution:',x.round(6))
print('np.solve:',np.linalg.solve(A,b).round(6))
```

```python
# Convergence plot
import matplotlib.pyplot as plt
A=np.array([[10.,-1.,2.],[-1.,11.,-1.],[2.,-1.,10.]])
b=np.array([6.,25.,-11.]); x=np.zeros(3); errs=[]
for _ in range(100):
    xn=np.linalg.solve(np.diag(np.diag(A)),b-(A-np.diag(np.diag(A)))@x)
    errs.append(np.linalg.norm(xn-x)); x=xn
plt.semilogy(errs); plt.xlabel('iter'); plt.ylabel('||Δx||'); plt.title('Jacobi'); plt.show()
```
