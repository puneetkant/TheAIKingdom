# 2.1.2 Matrices

Matrix operations for ML: OLS, covariance, broadcasting, softmax weight matrix.

```python
import numpy as np, matplotlib.pyplot as plt, urllib.request, csv
from pathlib import Path
DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'cal_housing.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)
with open(dest) as f: rows=list(csv.DictReader(f))
feat=['MedInc','HouseAge','AveRooms','AveBedrms','Population']
X=np.array([[float(r[c]) for c in feat] for r in rows[:200]])
y=np.array([float(r['MedHouseVal']) for r in rows[:200]])
print('X:',X.shape,'y:',y.shape)
```

```python
# Basic operations
print('Transpose:',X.T.shape)
print('Row means:',X.mean(0).round(3))
print('Element-wise multiply first col:',( X[:,0]*2 )[:5].round(3))
```

```python
# OLS via normal equations
Xb=np.hstack([np.ones((len(X),1)),X])
w=np.linalg.solve(Xb.T@Xb, Xb.T@y)
yh=Xb@w; ss_r=np.sum((y-yh)**2); ss_t=np.sum((y-y.mean())**2)
print(f'R² = {1-ss_r/ss_t:.4f}'); print('w =',w.round(4))
```

```python
# Covariance + correlation heatmap
Xc=X-X.mean(0); cov=(Xc.T@Xc)/(len(X)-1)
std=np.sqrt(np.diag(cov)); corr=cov/np.outer(std,std)
fig,ax=plt.subplots(figsize=(5,4))
im=ax.imshow(corr,cmap='coolwarm',vmin=-1,vmax=1); plt.colorbar(im)
ax.set_title('Feature Correlation'); plt.show()
```

```python
# Broadcasting: standardise without loop
X_std=(X-X.mean(0))/X.std(0)
print('Std mean:',X_std.mean(0).round(4))
print('Std std: ',X_std.std(0).round(4))
```
