# 2.1.4 Determinants

Properties, geometric area interpretation, invertibility, 3×3.

```python
import numpy as np, matplotlib.pyplot as plt
A=np.array([[2.,1.],[5.,3.]])
B=np.array([[1.,2.],[3.,4.]])
print('det(A)=',np.linalg.det(A))
print('det(AB)=',np.linalg.det(A@B),'= det(A)*det(B)=',np.linalg.det(A)*np.linalg.det(B))
```

```python
# Invertibility
for name,M in [('Full rank',np.array([[1.,2.],[3.,4.]])),('Singular',np.array([[1.,2.],[2.,4.]]))]:
    d=np.linalg.det(M)
    print(f'{name}: det={d:.4f}  {"invertible" if abs(d)>1e-10 else "SINGULAR"}')
```

```python
# Geometric: parallelogram area
a=np.array([3.,0.]); b=np.array([1.,2.])
area=abs(np.linalg.det(np.column_stack([a,b])))
print(f'Area = |det| = {area}')
corners=np.array([[0,0],a,a+b,b,[0,0]])
fig,ax=plt.subplots(figsize=(4,4))
ax.fill(corners[:,0],corners[:,1],alpha=0.3,color='steelblue')
for v,c,l in [(a,'b','a'),(b,'r','b')]: ax.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color=c,label=l)
ax.set_xlim(-0.5,5);ax.set_ylim(-0.5,3);ax.set_aspect('equal');ax.legend();ax.set_title(f'Area={area}');plt.show()
```

```python
# 3×3
A=np.array([[1.,2.,3.],[4.,5.,6.],[7.,8.,10.]])
print('det(A)=',np.linalg.det(A).round(4),'rank=',np.linalg.matrix_rank(A))
```
