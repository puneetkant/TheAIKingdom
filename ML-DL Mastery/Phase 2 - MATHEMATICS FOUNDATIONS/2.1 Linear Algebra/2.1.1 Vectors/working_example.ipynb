# 2.1.1 Vectors

ML-focused vector operations: norms, cosine similarity, projections, SGD gradient vectors.

```python
import numpy as np, matplotlib.pyplot as plt
a=np.array([1.,2.,3.]); b=np.array([4.,5.,6.])
print('a+b =',a+b)
print('dot(a,b)=',np.dot(a,b))
print('|a|₂ =',np.linalg.norm(a))
print('|a|₁ =',np.linalg.norm(a,1))
print('cos(a,b)=',np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b)))
```

```python
# Download Cal Housing
import urllib.request; from pathlib import Path
DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'cal_housing.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)
import csv
with open(dest) as f:
    rows=list(csv.DictReader(f))
feat=['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup']
X=np.array([[float(r[c]) for c in feat] for r in rows])
print('X shape:',X.shape)
```

```python
# Normalisation
norms=np.linalg.norm(X,axis=1,keepdims=True)
X_unit=X/(norms+1e-9)
print('L2 unit norm check:',np.linalg.norm(X_unit,axis=1).mean().round(6))
X_mm=(X-X.min(0))/(X.max(0)-X.min(0)+1e-9)
print('MinMax cols range:',X_mm.min(0).round(3),'→',X_mm.max(0).round(3))
```

```python
# Projection
a=np.array([3.,0.]); b=np.array([2.,4.])
proj=(np.dot(a,b)/np.dot(a,a))*a; rej=b-proj
fig,ax=plt.subplots(figsize=(5,5))
for v,c,lbl in [(a,'b','a'),(b,'r','b'),(proj,'g','proj'),(rej,'purple','rej')]:
    ax.quiver(0,0,v[0],v[1],angles='xy',scale_units='xy',scale=1,color=c,label=lbl,width=0.02)
ax.set_xlim(-1,5);ax.set_ylim(-1,5);ax.set_aspect('equal');ax.grid(0.3);ax.legend();ax.set_title('Projection')
plt.show()
```

```python
# SGD gradient demo
y=np.array([float(r['MedHouseVal']) for r in rows])
w=np.zeros(X.shape[1]); lr=0.001; losses=[]
for _ in range(50):
    err=X@w-y; losses.append(float(np.mean(err**2)))
    w-=lr*(X.T@err/len(y))
plt.plot(losses);plt.xlabel('Epoch');plt.ylabel('MSE');plt.title('SGD Loss');plt.grid(0.3);plt.show()
```
