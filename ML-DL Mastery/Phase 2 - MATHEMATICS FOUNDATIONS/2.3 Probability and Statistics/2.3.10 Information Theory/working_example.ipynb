# 2.3.10 Information Theory

Shannon entropy, KL divergence, cross-entropy loss, mutual information.

```python
import numpy as np, math
def H(p, base=2):
    p=np.array(p,dtype=float); p/=p.sum(); m=p>0
    return -np.sum(p[m]*(np.log2(p[m]) if base==2 else np.log(p[m])))
for p in [0.1,0.3,0.5,0.9]: print(f'H(Bernoulli({p}))={H([p,1-p]):.4f} bits')
print(f'H(fair die)={H([1/6]*6):.4f} bits (=log2(6)={math.log2(6):.4f})')
```

```python
# KL divergence
def kl(p,q):
    p,q=np.array(p,dtype=float),np.array(q,dtype=float)
    m=p>0; return np.sum(p[m]*np.log(p[m]/q[m]))
P=[.4,.3,.2,.1]; Q=[.35,.3,.25,.1]; Qf=[.1,.1,.1,.7]
print(f'KL(P||Q_close)={kl(P,Q):.4f}'); print(f'KL(P||Q_far)={kl(P,Qf):.4f}')
```

```python
# Cross-entropy (classification loss)
def softmax(z): e=np.exp(z-z.max()); return e/e.sum()
def ce(ytrue, probs): return -math.log(probs[ytrue]+1e-15)
for name,logits in [('good',[5.,1.,.5]),('bad',[.5,.5,.5])]:
    p=softmax(np.array(logits)); print(f'{name}: CE={ce(0,p):.4f}  probs={p.round(3)}')
```

```python
# Binary entropy plot
import matplotlib.pyplot as plt
ps=np.linspace(.01,.99,200)
hs=[-p*math.log2(p)-(1-p)*math.log2(1-p) for p in ps]
plt.plot(ps,hs); plt.xlabel('p'); plt.ylabel('H(p) bits'); plt.title('Binary entropy'); plt.show()
```
