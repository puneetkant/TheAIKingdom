# 2.3.7 Estimation Theory

MLE (Gaussian, Bernoulli), MAP with Beta prior, bias-variance decomposition.

```python
import numpy as np
np.random.seed(42); data=np.random.normal(3.,1.5,500)
mu_hat=data.mean(); sig_hat=np.sqrt(np.mean((data-mu_hat)**2))
print(f'MLE: μ={mu_hat:.4f} σ={sig_hat:.4f} (true: 3, 1.5)')
print(f'Unbiased σ: {np.std(data,ddof=1):.4f}')
```

```python
# MLE Bernoulli + CI
import math
for n in [10,50,200,1000]:
    d=np.random.binomial(1,.35,n); p=d.mean()
    se=math.sqrt(p*(1-p)/n)
    print(f'n={n:>5}: p̂={p:.4f} CI=[{max(0,p-1.96*se):.4f},{min(1,p+1.96*se):.4f}]')
```

```python
# MAP with Beta conjugate prior
alpha,beta,n,k=2,2,20,4
print('MLE:', k/n, 'MAP:', (alpha+k-1)/(alpha+beta+n-2), 'Post mean:', (alpha+k)/(alpha+beta+n))
```

```python
# Bias-Variance decomp (poly regression)
np.random.seed(3); x=np.linspace(0,2*3.14159,15); xt=np.linspace(0,2*3.14159,200)
preds={k:[] for k in [1,3,9]}
for _ in range(200):
    y=np.sin(x)+np.random.normal(0,.3,15)
    for k in [1,3,9]: preds[k].append(np.polyval(np.polyfit(x,y,k),xt))
yt=np.sin(xt)
print('Degree  Bias²   Var    MSE')
for k in [1,3,9]:
    p=np.array(preds[k]); mp=p.mean(0)
    print(f'  {k}  {np.mean((mp-yt)**2):.4f}  {np.mean(p.var(0)):.4f}  {np.mean((p-yt)**2):.4f}')
```
