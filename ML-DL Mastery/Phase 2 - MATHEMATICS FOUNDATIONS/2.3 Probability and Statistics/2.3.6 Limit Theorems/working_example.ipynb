# 2.3.6 Limit Theorems

Law of Large Numbers, Central Limit Theorem, bootstrap CI, Monte Carlo integration.

```python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(42); p=0.6; s=np.random.binomial(1,p,10000)
running=np.cumsum(s)/np.arange(1,10001)
plt.semilogx(running,label='running mean'); plt.axhline(p,color='r',ls='--',label=f'true p={p}')
plt.xlabel('n'); plt.legend(); plt.title('LLN'); plt.show()
```

```python
# CLT
np.random.seed(0); fig,axes=plt.subplots(1,4,figsize=(12,3))
for ax,n in zip(axes,[1,5,30,100]):
    means=np.random.exponential(1,(10000,n)).mean(1)
    z=(means-1)/(1/np.sqrt(n))
    ax.hist(z,bins=50,density=True,alpha=0.6,color='coral')
    x=np.linspace(-4,4,200); ax.plot(x,np.exp(-x**2/2)/np.sqrt(2*np.pi),'k')
    ax.set_title(f'n={n}'); ax.set_xlim(-4,4)
plt.tight_layout(); plt.show()
```

```python
# Bootstrap CI
np.random.seed(7); data=np.random.exponential(2.,50)
boot=[np.random.choice(data,50,replace=True).mean() for _ in range(10000)]
ci=np.percentile(boot,[2.5,97.5])
print(f'Obs mean={data.mean():.4f}  95%CI=[{ci[0]:.4f},{ci[1]:.4f}]  in CI:{ci[0]<=2<=ci[1]}')
```

```python
# MC integration: ∫₀¹ x² dx = 1/3
for N in [100,1000,100000]:
    x=np.random.uniform(0,1,N); est=x**2
    print(f'N={N}: {np.mean(est):.5f} ±{1.96*np.std(est)/N**0.5:.5f} (true=0.33333)')
```
