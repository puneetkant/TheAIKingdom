# 2.3.8 Hypothesis Testing

One-sample t-test, A/B two-sample t-test, chi-squared, Type I/II errors.

```python
import numpy as np, math
np.random.seed(42); data=np.random.normal(5.3,1.5,30)
def ttest1(d,mu0):
    n=len(d); t=(d.mean()-mu0)/(d.std(ddof=1)/n**0.5)
    p=2*(1-0.5*(1+math.erf(abs(t)/math.sqrt(2)))); return t,p
for mu0 in [5.,5.3,6.]:
    t,p=ttest1(data,mu0)
    print(f'H0:μ={mu0}: t={t:.3f} p={p:.4f} {"REJECT" if p<.05 else "FAIL TO REJECT"}')
```

```python
# Two-sample A/B test
np.random.seed(1); ctrl=np.random.normal(10.,2.,50); trt=np.random.normal(12.,2.,50)
def ttest2(a,b):
    n1,n2=len(a),len(b)
    sp=math.sqrt(((n1-1)*a.var(ddof=1)+(n2-1)*b.var(ddof=1))/(n1+n2-2))
    t=(a.mean()-b.mean())/(sp*(1/n1+1/n2)**.5)
    p=2*(1-0.5*(1+math.erf(abs(t)/math.sqrt(2)))); return t,p
t,p=ttest2(ctrl,trt); print(f't={t:.3f} p={p:.4f} {"SIGNIFICANT" if p<.05 else "NOT SIG"}')
```

```python
# Chi-squared test
obs=np.array([[120,880],[160,840]])
n=obs.sum(); rs=obs.sum(1,keepdims=True); cs=obs.sum(0,keepdims=True)
exp=rs@cs/n; chi2=np.sum((obs-exp)**2/exp)
print(f'chi2={chi2:.4f} critical=3.841 → {"SIG" if chi2>3.841 else "NOT SIG"}')
```

```python
# Type I and II errors
np.random.seed(0); n=30
t1=sum(abs(np.random.normal(0,1,n).mean())>1.96/n**0.5 for _ in range(10000))/10000
t2=sum(abs(np.random.normal(.5,1,n).mean())<=1.96/n**0.5 for _ in range(10000))/10000
print(f'Type I (α)={t1:.4f} Type II (β)={t2:.4f} Power={1-t2:.4f}')
```
