# 2.3.3 Random Variables

Discrete (Binomial, Poisson) and continuous (Gaussian) RVs; moments, PDF, CDF.

```python
import numpy as np
np.random.seed(42)
# Binomial
s=np.random.binomial(10,0.3,10000)
print(f'Binomial E={np.mean(s):.3f} Var={np.var(s):.3f} (theory: E=3, Var=2.1)')
# Poisson
s=np.random.poisson(3,10000)
print(f'Poisson  E={np.mean(s):.3f} Var={np.var(s):.3f} (theory: E=Var=3)')
```

```python
import matplotlib.pyplot as plt, math
mu,sig=2.,1.5; s=np.random.normal(mu,sig,50000)
print(f'Gaussian E={np.mean(s):.3f} std={np.std(s):.3f}')
for t in [0.,mu-sig,mu,mu+sig]: print(f'P(X<={t:.1f})={np.mean(s<=t):.4f}')
```

```python
# PDF and CDF plots
mu,sig=2.,1.5
s=np.random.normal(mu,sig,50000)
x=np.linspace(mu-4*sig,mu+4*sig,300)
pdf=np.exp(-0.5*((x-mu)/sig)**2)/(sig*(2*math.pi)**0.5)
fig,ax=plt.subplots(figsize=(6,4))
ax.hist(s,bins=80,density=True,alpha=0.4)
ax.plot(x,pdf,'r',lw=2,label='PDF'); ax.legend(); plt.show()
```

```python
# Moments of Exponential
s=np.random.exponential(2.,100000)
print('E:',np.mean(s),'Var:',np.var(s),'Skew:',np.mean(((s-s.mean())/s.std())**3))
```
