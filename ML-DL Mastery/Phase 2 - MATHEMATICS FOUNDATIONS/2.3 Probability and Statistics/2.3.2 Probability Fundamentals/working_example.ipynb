# 2.3.2 Probability Fundamentals

Sample spaces, Bayes theorem, conditional independence, Monte Carlo π.

```python
import random; random.seed(99)
rolls=[random.randint(1,6) for _ in range(100000)]
for f in range(1,7): print(f'P(die={f})≈{rolls.count(f)/100000:.4f}')
```

```python
# Bayes: P(disease|positive test)
PD=0.001; P_pos_D=0.99; P_pos_nD=0.01
P_pos=P_pos_D*PD+P_pos_nD*(1-PD)
P_D_pos=P_pos_D*PD/P_pos
print(f'P(D|+)={P_D_pos:.4f}  only {P_D_pos*100:.1f}% even with 99% accurate test!')
```

```python
# Conditional independence
random.seed(0); N=100000
Z=[random.random()<0.5 for _ in range(N)]
X=[random.random()<(0.7 if z else 0.3) for z in Z]
Y=[random.random()<(0.7 if z else 0.3) for z in Z]
print('P(X,Y)=',sum(a and b for a,b in zip(X,Y))/N)
print('P(X)P(Y)=',sum(X)/N*sum(Y)/N)
```

```python
import math
for N in [1000,10000,1000000]:
    inside=sum(random.random()**2+random.random()**2<1 for _ in range(N))
    pi=4*inside/N
    print(f'N={N}: π≈{pi:.5f}  err={abs(pi-math.pi):.5f}')
```
