# 2.3.5 Joint Distributions and Transformations

Bivariate Gaussian, covariance/correlation, Box-Muller transform, marginalisation.

```python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(42)
mu=np.array([1.,-2.]); sigma=np.array([[2.,.8],[.8,1.]])
L=np.linalg.cholesky(sigma); X=np.random.randn(1000,2)@L.T+mu
print('mean:',X.mean(0).round(4))
print('cov:\n',np.cov(X.T).round(4))
plt.scatter(X[:,0],X[:,1],s=4,alpha=0.3); plt.title('Bivariate Gaussian'); plt.show()
```

```python
# Marginalisation
plt.hist(X[:,0],bins=40,density=True,alpha=0.6,label='X1')
plt.hist(X[:,1],bins=40,density=True,alpha=0.6,label='X2')
plt.legend(); plt.title('Marginals'); plt.show()
print('corr:',np.corrcoef(X.T)[0,1].round(4))
```

```python
# Box-Muller: Uniform -> Gaussian
U1=np.random.uniform(0,1,50000); U2=np.random.uniform(0,1,50000)
Z1=np.sqrt(-2*np.log(U1))*np.cos(2*np.pi*U2)
print('Z1 mean:',Z1.mean().round(4),'std:',Z1.std().round(4))
```

```python
# Correlation heatmap
import csv,urllib.request
from pathlib import Path
dest=Path('data/cal_housing.csv'); dest.parent.mkdir(exist_ok=True)
if not dest.exists(): urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)
with open(dest) as f: rows=list(csv.DictReader(f))
feat=['MedInc','HouseAge','AveRooms','AveBedrms']
X=np.array([[float(r[c]) for c in feat] for r in rows[:300]])
corr=np.corrcoef(X.T)
fig,ax=plt.subplots(figsize=(5,4))
im=ax.imshow(corr,cmap='RdBu_r',vmin=-1,vmax=1)
ax.set_xticks(range(4)); ax.set_xticklabels(feat,rotation=30,ha='right')
ax.set_yticks(range(4)); ax.set_yticklabels(feat)
plt.colorbar(im); ax.set_title('Correlation'); plt.show()
```
