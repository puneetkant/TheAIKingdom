# 2.3.11 Stochastic Processes (Fundamentals)

Markov chains, stationary distribution, random walk, Brownian motion, Geometric BM.

```python
import numpy as np
np.random.seed(0)
P=np.array([[.7,.2,.1],[.3,.4,.3],[.2,.3,.5]])
vals,vecs=np.linalg.eig(P.T); idx=np.argmin(np.abs(vals-1))
stat=np.real(vecs[:,idx]); stat/=stat.sum()
print('Stationary:',stat.round(4))
print('P^100 row0:',np.linalg.matrix_power(P,100)[0].round(4))
```

```python
# Random walk
import matplotlib.pyplot as plt
np.random.seed(1); n=1000
steps=np.random.choice([-1,1],size=(20,n)); paths=np.cumsum(steps,axis=1)
for p in paths[:5]: plt.plot(p,lw=.6,alpha=.7)
plt.fill_between(range(n),-np.sqrt(np.arange(n)),np.sqrt(np.arange(n)),alpha=.15,color='grey')
plt.title('Random walks'); plt.show()
```

```python
# Brownian motion
np.random.seed(2); T,n=1.,1000; dt=T/n; t=np.linspace(0,T,n+1)
fig,axes=plt.subplots(1,2,figsize=(11,4))
for _ in range(5):
    W=np.concatenate([[0],np.cumsum(np.random.normal(0,dt**.5,n))])
    axes[0].plot(t,W,lw=.7,alpha=.7)
axes[0].set_title('Brownian motion')
S0,mu,sig=100,.1,.2
for _ in range(5):
    W=np.cumsum(np.random.normal(0,dt**.5,n))
    S=S0*np.exp((mu-.5*sig**2)*t[1:]+sig*W)
    axes[1].plot(t[1:],S,lw=.7,alpha=.7)
axes[1].set_title('Geometric BM'); plt.show()
```

```python
# Empirical stationary dist vs analytic
np.random.seed(0); state=0; counts=[0,0,0]
for _ in range(10000):
    state=np.random.choice(3,p=P[state]); counts[state]+=1
emp=np.array(counts)/10000
print(f'Empirical : {emp.round(4)}')
print(f'Analytic  : {stat.round(4)}')
```
