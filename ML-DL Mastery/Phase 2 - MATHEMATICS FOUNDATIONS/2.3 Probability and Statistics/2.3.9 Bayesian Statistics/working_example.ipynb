# 2.3.9 Bayesian Statistics

Beta-Binomial conjugate update, sequential Bayesian learning, Metropolis-Hastings MCMC.

```python
import numpy as np, math
np.random.seed(42); data=np.random.binomial(1,.3,100)
k,n=data.sum(),len(data); a,b=2+k,2+n-k
print(f'Prior mean={2/4:.3f}  Posterior mean={a/(a+b):.3f}  MLE={k/n:.3f}')
```

```python
import matplotlib.pyplot as plt
def beta_pdf(x,a,b):
    lB=math.lgamma(a)+math.lgamma(b)-math.lgamma(a+b)
    return np.exp((a-1)*np.log(x)+(b-1)*np.log(1-x)-lB)
x=np.linspace(0.001,.999,300)
plt.plot(x,beta_pdf(x,2,2),'--',label='Prior Beta(2,2)')
plt.plot(x,beta_pdf(x,a,b),lw=2,label=f'Posterior Beta({a},{b})')
plt.axvline(.3,ls=':',label='true p'); plt.legend(); plt.title('Bayesian update'); plt.show()
```

```python
# Sequential update
alpha,beta_=1,1; p_true=0.4; np.random.seed(0)
flips=np.random.binomial(1,p_true,50)
for i,x in enumerate(flips,1):
    alpha+=x; beta_+=(1-x)
    if i in [1,5,10,25,50]:
        pm=alpha/(alpha+beta_)
        print(f'n={i:>3}: post_mean={pm:.4f}')
```

```python
# Metropolis-Hastings
import random; random.seed(7); np.random.seed(7)
data=np.random.normal(2.5,1.,20)
def log_post(mu): return -0.5*np.sum((data-mu)**2)-0.5*(mu/3)**2
mu=0.; samples=[]
for _ in range(20000):
    prop=mu+np.random.normal(0,.5)
    if math.log(random.random()+1e-15)<log_post(prop)-log_post(mu): mu=prop
    samples.append(mu)
s=np.array(samples[5000:])
print(f'MCMC mean={s.mean():.4f} std={s.std():.4f}')
plt.hist(s,bins=60,density=True); plt.title('MCMC posterior'); plt.show()
```
