# 2.3.1 Combinatorics

Permutations, combinations, birthday problem Monte Carlo, multinomial, entropy.

```python
import math
n,k=10,3
print('P(n,k):',math.perm(n,k),'  C(n,k):',math.comb(n,k))
print('Pascal row 5:',[math.comb(5,i) for i in range(6)])
```

```python
# Birthday problem
import random; random.seed(42)
def prob(n,T=50000):
    return sum(len({random.randint(0,364) for _ in range(n)})<n for _ in range(T))/T
for n in [10,20,23,30,50]:
    ana=1-math.prod((365-i)/365 for i in range(1,n))
    print(f'n={n:>2}: MC={prob(n):.4f}  analytic={ana:.4f}')
```

```python
from collections import Counter
word='MISSISSIPPI'; c=Counter(word)
ans=math.factorial(len(word))//math.prod(math.factorial(v) for v in c.values())
print(f'Arrangements of {word}: {ans:,}')
```
