# 2.3.4 Common Probability Distributions

Gaussian, Laplace, Beta, Binomial, Poisson, Student-t, Chi², Dirichlet.

```python
import numpy as np, matplotlib.pyplot as plt, math
np.random.seed(42)
x=np.linspace(-5,5,300)
fig,ax=plt.subplots(figsize=(8,4))
for mu,sig in [(0,1),(0,2),(2,0.5)]:
    y=np.exp(-0.5*((x-mu)/sig)**2)/(sig*math.sqrt(2*math.pi))
    ax.plot(x,y,label=f'μ={mu},σ={sig}')
ax.legend(); ax.set_title('Gaussian'); plt.show()
```

```python
# Laplace vs Gaussian tails
fig,ax=plt.subplots(figsize=(8,4))
ax.semilogy(x,np.exp(-0.5*x**2)/math.sqrt(2*math.pi),label='Gaussian')
ax.semilogy(x,np.exp(-np.abs(x))/2,label='Laplace')
ax.legend(); ax.set_title('Heavy tails'); plt.show()
print('Laplace: heavier tails → L1/Lasso regularisation')
```

```python
# Beta distribution
def beta_pdf(x,a,b):
    B=math.lgamma(a)+math.lgamma(b)-math.lgamma(a+b)
    return np.exp((a-1)*np.log(x)+(b-1)*np.log(1-x)-B)
xb=np.linspace(0.001,0.999,300)
fig,ax=plt.subplots(figsize=(7,4))
for a,b in [(2,2),(5,2),(1,1),(0.5,0.5)]:
    ax.plot(xb,beta_pdf(xb,a,b),label=f'α={a},β={b}')
ax.legend(); ax.set_title('Beta distribution — prior for p'); plt.show()
```

```python
# Moments summary
for name,s in [('Gaussian',np.random.normal(0,1,100000)),('Laplace',np.random.laplace(0,1,100000)),
               ('Exp(1)',np.random.exponential(1,100000)),('t(5)',np.random.standard_t(5,100000))]:
    m=s.mean(); v=s.var(); std=s.std()
    print(f'{name:15s} E={m:.3f} Var={v:.3f} Skew={np.mean(((s-m)/std)**3):.3f} Kurt={np.mean(((s-m)/std)**4)-3:.3f}')
```
