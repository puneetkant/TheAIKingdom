# 2.2.4 Optimization Theory

Convexity, GD, SGD, Momentum, Adam, Newton's method — from scratch.

```python
import numpy as np
# Convexity: check Hessian PSD
H=np.array([[2.,1.],[1.,4.]]); eigs=np.linalg.eigvalsh(H)
print('f=x²+xy+2y², eigs:',eigs,'→',('CONVEX' if all(eigs>=0) else 'NOT CONVEX'))
H2=np.array([[2.,0.],[0.,-2.]]); print('f=x²-y², eigs:',np.linalg.eigvalsh(H2),'→ SADDLE')
```

```python
import matplotlib.pyplot as plt
np.random.seed(0); n,d=100,10; lam=0.01
X=np.random.randn(n,d); y=X@np.random.randn(d)+0.1*np.random.randn(n)
full_g=lambda w: 2*X.T@(X@w-y)/n+2*lam*w
loss=lambda w: np.mean((X@w-y)**2)+lam*np.dot(w,w)
# GD
w=np.zeros(d); gd=[]
for _ in range(200): w-=0.1*full_g(w); gd.append(loss(w))
# Adam
w=np.zeros(d); m=np.zeros(d); v=np.zeros(d); adam=[]
for t in range(1,201):
    idx=np.random.choice(n,16); Xi=X[idx]; yi=y[idx]
    g=2*Xi.T@(Xi@w-yi)/16+2*lam*w
    m=0.9*m+0.1*g; v=0.999*v+0.001*g**2
    w-=0.05*(m/(1-0.9**t))/(np.sqrt(v/(1-0.999**t))+1e-8); adam.append(loss(w))
plt.semilogy(gd,label='GD'); plt.semilogy(adam,label='Adam')
plt.xlabel('iter'); plt.ylabel('loss'); plt.legend(); plt.show()
```

```python
# Newton's method
f=lambda x: x**4-4*x**2+x; df=lambda x: 4*x**3-8*x+1; d2f=lambda x: 12*x**2-8
x=2.; path=[x]
for _ in range(10): x-=df(x)/d2f(x); path.append(x)
print(f'Newton min: x={x:.6f} f={f(x):.6f}')
xs=np.linspace(-2.5,2.5,300); plt.plot(xs,f(xs)); plt.scatter(path,[f(p) for p in path],c='r'); plt.ylim(-10,10); plt.show()
```
