# 2.2.2 Multivariable Calculus

Numerical gradient, Jacobian, Hessian, gradient descent on Rosenbrock function.

```python
import numpy as np
def grad(f,x,h=1e-6):
    g=np.zeros_like(x,dtype=float)
    for i in range(len(x)):
        xp,xm=x.copy(),x.copy(); xp[i]+=h; xm[i]-=h
        g[i]=(f(xp)-f(xm))/(2*h)
    return g
f=lambda v: v[0]**2+2*v[1]**2+v[0]*v[1]
x=np.array([1.,2.])
print('gradient:',grad(f,x))
print('analytic:',[2*x[0]+x[1],4*x[1]+x[0]])
```

```python
# Hessian via finite diff
def hess(f,x,h=1e-5):
    n=len(x); H=np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            xij,xi,xj=x.copy(),x.copy(),x.copy()
            xij[i]+=h;xij[j]+=h;xi[i]+=h;xj[j]+=h
            H[i,j]=(f(xij)-f(xi)-f(xj)+f(x))/h**2
    return H
H=hess(f,x); print('H:\n',H.round(4))
print('eigenvalues:',np.linalg.eigvalsh(H))
```

```python
# GD on Rosenbrock
f=lambda v:(1-v[0])**2+100*(v[1]-v[0]**2)**2
x=np.array([-1.,1.]); lr=1e-3; path=[x.copy()]
for _ in range(3000): g=grad(f,x); x=x-lr*g; path.append(x.copy())
print('min:',x.round(4))
```

```python
import matplotlib.pyplot as plt
path=np.array(path)
xr=np.linspace(-2,2,200); yr=np.linspace(-1,3,200)
XX,YY=np.meshgrid(xr,yr); ZZ=(1-XX)**2+100*(YY-XX**2)**2
fig,ax=plt.subplots(figsize=(7,5))
ax.contour(XX,YY,np.log1p(ZZ),30,cmap='viridis')
ax.plot(path[:,0],path[:,1],'r-',lw=0.7)
ax.scatter(1,1,c='g',zorder=5,label='true min'); ax.legend(); plt.show()
```
