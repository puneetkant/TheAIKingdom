# 2.2.1 Single Variable Calculus

Numerical derivatives, chain rule, Taylor series, 1D gradient descent.

```python
import numpy as np
# Central difference numerical derivative
f=lambda x: np.sin(x)+0.5*x**2; df=lambda x: np.cos(x)+x; x0=1.0
for h in [1e-1,1e-4,1e-7,1e-10]:
    d=(f(x0+h)-f(x0-h))/(2*h)
    print(f'h={h:.0e}: {d:.8f}  err={abs(d-df(x0)):.2e}')
```

```python
# Chain rule: d/dx sigmoid(x^2) = sig(x^2)*(1-sig(x^2))*2x
from math import exp
def sig(u): return 1/(1+exp(-u))
f=lambda x: sig(x**2); df=lambda x: sig(x**2)*(1-sig(x**2))*2*x
for x in [0.5,1.0,2.0]:
    h=1e-7; num=(f(x+h)-f(x-h))/(2*h)
    print(f'x={x}: analytic={df(x):.6f}  numeric={num:.6f}  err={abs(df(x)-num):.2e}')
```

```python
import matplotlib.pyplot as plt, numpy as np
x=np.linspace(-np.pi,np.pi,300)
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(x,np.sin(x),'k',lw=2,label='sin(x)')
ax.plot(x,x,'--b',label='T1'); ax.plot(x,x-x**3/6,'--g',label='T3')
ax.plot(x,x-x**3/6+x**5/120,'--r',label='T5')
ax.set_ylim(-2,2); ax.legend(); ax.set_title('Taylor series'); plt.show()
```

```python
# 1D gradient descent: min (x-3)^2+2
f=lambda x:(x-3)**2+2; df=lambda x:2*(x-3)
x=0.; lr=0.1; path=[x]
for _ in range(50): x=x-lr*df(x); path.append(x)
print(f'x*={path[-1]:.6f}  f(x*)={f(path[-1]):.6f}')
plt.semilogy([abs(p-3) for p in path]); plt.xlabel('iter'); plt.ylabel('|x-3|'); plt.show()
```
