# 2.2.3 Matrix Calculus

Numerically verify: d/dw‖Xw-y‖², softmax Jacobian, Ridge gradient.

```python
import numpy as np
np.random.seed(42); n,d=20,5
X=np.random.randn(n,d); y=np.random.randn(n); w=np.random.randn(d)
f=lambda v: np.sum((X@v-y)**2)
g_ana=2*X.T@(X@w-y)
h=1e-7
g_num=np.array([(f(w+h*np.eye(d)[i])-f(w-h*np.eye(d)[i]))/(2*h) for i in range(d)])
print('ana:',g_ana.round(4)); print('num:',g_num.round(4))
print('err:',np.max(np.abs(g_ana-g_num)))
```

```python
# Softmax Jacobian
def sm(z): e=np.exp(z-z.max()); return e/e.sum()
z=np.array([2.,1.,0.1]); s=sm(z)
J_ana=np.diag(s)-np.outer(s,s)
h=1e-7
J_num=np.array([(sm(z+h*np.eye(3)[i])-sm(z-h*np.eye(3)[i]))/(2*h) for i in range(3)]).T
print('J_ana:\n',J_ana.round(4))
print('max err:',np.max(np.abs(J_ana-J_num)))
```

```python
# Ridge regression GD
import matplotlib.pyplot as plt
np.random.seed(1); n,d=50,8; lam=0.1
X=np.random.randn(n,d); y=X@np.ones(d)+0.1*np.random.randn(n)
w=np.zeros(d); losses=[]
for _ in range(500):
    g=2*X.T@(X@w-y)+2*lam*w; w-=0.01*g
    losses.append(np.sum((X@w-y)**2)+lam*np.dot(w,w))
w_opt=np.linalg.solve(2*(X.T@X+lam*np.eye(d)),2*X.T@y)
print('GD w:',w.round(4)); print('opt w:',w_opt.round(4))
plt.semilogy(losses); plt.xlabel('iter'); plt.ylabel('loss'); plt.show()
```
