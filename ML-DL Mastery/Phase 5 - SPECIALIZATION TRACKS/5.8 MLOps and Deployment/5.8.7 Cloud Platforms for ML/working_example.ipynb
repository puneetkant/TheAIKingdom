# 5.8.7 Cloud Platforms for ML

AWS SageMaker, GCP Vertex AI, Azure ML — managed training, serving, cost estimation.

```python
# Cloud cost estimator
prices = {'aws_p3.2xl':3.06,'gcp_a2_1g':3.67,'azure_nc6s':3.06,'aws_g4dn':0.53}
hours = 20
print('Training cost estimates (20h):')
for name,p in prices.items(): print(f'  {name:20s} on-demand=${p*hours:.2f}  spot≈${p*hours*0.3:.2f}')
```

```python
import numpy as np, matplotlib.pyplot as plt
hs=np.arange(1,51)
plt.plot(hs,hs*3.06,label='AWS p3 (on-demand)')
plt.plot(hs,hs*3.06*0.3,label='AWS p3 (spot)')
plt.plot(hs,hs*0.53,label='AWS g4dn (inference)')
plt.xlabel('Hours'); plt.ylabel('USD'); plt.legend(); plt.title('Cloud Cost'); plt.show()
```
