# 5.8.5 CI/CD for ML

Automated pipelines: lint → test → train → evaluate → deploy.

```python
import numpy as np
# Simple training stage
def train(seed=0):
    np.random.seed(seed); N=150; X=np.random.randn(N,3); y=(X[:,0]>0).astype(int)
    W=np.zeros(4); Xb=np.column_stack([np.ones(N),X])
    for _ in range(80):
        p=1/(1+np.exp(-(Xb@W).clip(-50,50))); W-=0.2*(Xb.T@(p-y))/N
    preds=(1/(1+np.exp(-(Xb@W).clip(-50,50)))>=0.5).astype(int)
    return W, (preds==y).mean()
W, acc = train(); print(f'val_acc={acc:.4f}')
```

```python
# Eval gate
threshold=0.70; passed=acc>=threshold
print('Gate:','PASS' if passed else 'FAIL',f'({acc:.4f} vs {threshold})')
# Show pipeline stages
import matplotlib.pyplot as plt
stages=['lint','test','train','eval','deploy']; passed_all=[True]*5
fig,ax=plt.subplots(figsize=(8,1.5))
for i,(s,p) in enumerate(zip(stages,passed_all)):
    ax.barh(0,1,left=i,color='green' if p else 'red',height=0.5,edgecolor='white')
    ax.text(i+0.5,0,s,ha='center',va='center',color='white',fontsize=9)
ax.axis('off'); plt.title('Pipeline'); plt.show()
```
