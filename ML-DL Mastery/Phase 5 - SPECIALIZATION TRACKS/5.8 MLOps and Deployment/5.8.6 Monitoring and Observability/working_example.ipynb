# 5.8.6 Monitoring and Observability

Data drift (KS, PSI), prediction drift, alerting, SLA tracking.

```python
import numpy as np
# KS drift detection
def ks_drift(ref, prod):
    n1,n2=len(ref),len(prod); vals=np.sort(np.concatenate([ref,prod]))
    c1=np.searchsorted(np.sort(ref),vals,'right')/n1
    c2=np.searchsorted(np.sort(prod),vals,'right')/n2
    stat=np.max(np.abs(c1-c2)); crit=1.36*np.sqrt((n1+n2)/(n1*n2))
    return stat, stat>crit
ref=np.random.randn(300); prod=np.random.randn(300)+0.5
stat,drifted=ks_drift(ref,prod)
print(f'KS={stat:.3f}  Drifted={drifted}')
```

```python
# PSI score
def psi(ref, prod, bins=10):
    rh,edges=np.histogram(ref,bins=bins); ph,_=np.histogram(prod,bins=edges)
    rp=(rh+1e-8)/rh.sum(); pp=(ph+1e-8)/ph.sum()
    return float(np.sum((pp-rp)*np.log(pp/rp)))
print(f'PSI={psi(ref,prod):.4f}  (>0.2 = significant drift)')
import matplotlib.pyplot as plt
plt.hist(ref,bins=30,alpha=0.5,label='reference'); plt.hist(prod,bins=30,alpha=0.5,label='production')
plt.legend(); plt.title('Feature Distribution Shift'); plt.show()
```
