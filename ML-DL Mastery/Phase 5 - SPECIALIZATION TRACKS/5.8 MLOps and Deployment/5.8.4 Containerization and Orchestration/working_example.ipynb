# 5.8.4 Containerisation and Orchestration

Docker, docker-compose, Kubernetes, resource limits, health checks.

```python
# Generate a minimal Dockerfile string
def dockerfile(port=8000):
    return f'''FROM python:3.12-slim\nWORKDIR /app\nCOPY requirements.txt .\nRUN pip install --no-cache-dir -r requirements.txt\nCOPY . .\nEXPOSE {port}\nCMD ["uvicorn","serve:app","--host","0.0.0.0","--port","{port}"]'''
print(dockerfile())
```
