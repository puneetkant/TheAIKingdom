Training pipeline, data validation, preprocessing, serving interface.

```python
import numpy as np
# Minimal feature normalisation pipeline
class Normalizer:
    def fit(self,X): self.m=X.mean(0); self.s=X.std(0).clip(1e-8); return self
    def transform(self,X): return (X-self.m)/self.s
    def fit_transform(self,X): return self.fit(X).transform(X)
np.random.seed(0); X=np.random.randn(100,4)
Xn=Normalizer().fit_transform(X)
print('Mean ~0:',Xn.mean(0).round(3),'  Std ~1:',Xn.std(0).round(3))
```

```python
# Data validation
def validate(X, y, F=4):
    assert X.shape[1]==F; assert not np.isnan(X).any()
    assert len(X)==len(y); print('Validation passed')
validate(X, np.random.randint(0,2,100))
```
