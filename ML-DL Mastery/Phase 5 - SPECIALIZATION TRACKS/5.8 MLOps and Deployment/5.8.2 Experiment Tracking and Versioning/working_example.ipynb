# 5.8.2 Experiment Tracking and Versioning

Run logging, hyperparameter sweeps, model registry, Git integration.

```python
import numpy as np
# Lightweight run tracker
runs=[]
def log_run(params, metrics):
    runs.append({'params':params,'metrics':metrics})
    print(f'  Run {len(runs)}: {params} → {metrics}')
for lr in [0.01,0.1,0.5]:
    acc=0.5+0.4*(1-np.exp(-lr*10))+np.random.randn()*0.02
    log_run({'lr':lr},{'val_acc':round(acc,4)})
```

```python
# Compare runs
best=max(runs,key=lambda r:r['metrics']['val_acc'])
print('Best:',best)
import matplotlib.pyplot as plt
lrs=[r['params']['lr'] for r in runs]; accs=[r['metrics']['val_acc'] for r in runs]
plt.plot(lrs,accs,'o-'); plt.xscale('log'); plt.title('LR sweep'); plt.show()
```
