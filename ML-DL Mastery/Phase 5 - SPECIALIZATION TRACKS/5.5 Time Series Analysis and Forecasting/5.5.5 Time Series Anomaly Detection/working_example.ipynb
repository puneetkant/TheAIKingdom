# 5.5.5 Time Series Anomaly Detection

Z-score, IQR, CUSUM, Isolation Forest, LSTM reconstruction error.

```python
import numpy as np
rng=np.random.default_rng(42); t=np.arange(300)
ts=np.sin(2*np.pi*t/30)*3+rng.normal(0,.2,300)
anomalies=[50,100,150]; ts[anomalies]+=6
print('Injected anomalies at indices:',anomalies)
```

```python
import matplotlib.pyplot as plt
# Rolling Z-score
W=20; scores=np.zeros(300)
for i in range(W,300):
    w=ts[i-W:i]; scores[i]=abs(ts[i]-w.mean())/(w.std()+1e-8)
flags=scores>3
plt.figure(figsize=(10,4)); plt.plot(ts,alpha=.6,label='Series')
plt.scatter(np.where(flags)[0],ts[flags],color='red',zorder=5,label='Detected')
plt.scatter(anomalies,ts[anomalies],marker='x',color='green',s=80,label='True')
plt.legend(); plt.title('Anomaly Detection (Rolling Z-Score)'); plt.show()
```
