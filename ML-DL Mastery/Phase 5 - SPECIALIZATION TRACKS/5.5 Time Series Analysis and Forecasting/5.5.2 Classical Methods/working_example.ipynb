# 5.5.2 Classical Methods

ARIMA, SARIMA, ETS, exponential smoothing, Holt-Winters.

```python
import numpy as np
# AR(1) simulation: x_t = phi * x_{t-1} + eps
phi=0.8; n=200; rng=np.random.default_rng(42)
x=np.zeros(n)
for t in range(1,n): x[t]=phi*x[t-1]+rng.normal()
print('AR(1) phi=0.8, sample mean:',x.mean().round(3),'  std:',x.std().round(3))
```

```python
import matplotlib.pyplot as plt
# Simple exponential smoothing
alpha=0.3; s=np.zeros(n); s[0]=x[0]
for t in range(1,n): s[t]=alpha*x[t]+(1-alpha)*s[t-1]
plt.plot(x,alpha=0.5,label='Observed'); plt.plot(s,label=f'SES α={alpha}')
plt.legend(); plt.title('Exponential Smoothing on AR(1)'); plt.show()
```
