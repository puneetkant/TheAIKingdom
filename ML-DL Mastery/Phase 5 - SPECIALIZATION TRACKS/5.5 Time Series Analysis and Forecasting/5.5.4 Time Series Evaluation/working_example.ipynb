# 5.5.4 Time Series Evaluation

MAE, RMSE, MAPE, sMAPE, backtesting, walk-forward validation.

```python
import numpy as np
def mae(y,yh): return np.mean(np.abs(y-yh))
def rmse(y,yh): return np.sqrt(np.mean((y-yh)**2))
def mape(y,yh): return np.mean(np.abs((y-yh)/(np.abs(y)+1e-8)))*100
n=200; t=np.arange(n); ts=np.sin(2*np.pi*t/20)*3+0.01*t+np.random.normal(0,.2,n)
train,test=ts[:160],ts[160:]; naive=np.full(len(test),train[-1])
print(f'Naive  MAE={mae(test,naive):.3f} RMSE={rmse(test,naive):.3f} MAPE={mape(test,naive):.2f}%')
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(160),train,label='Train'); plt.plot(np.arange(160,200),test,label='Test')
plt.plot(np.arange(160,200),naive,ls='--',label='Naive')
plt.legend(); plt.title('Naive Forecast vs Test'); plt.show()
```
