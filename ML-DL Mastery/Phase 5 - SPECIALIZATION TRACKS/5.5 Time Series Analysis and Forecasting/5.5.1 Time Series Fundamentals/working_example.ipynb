# 5.5.1 Time Series Fundamentals

Stationarity, ACF/PACF, trend, seasonality, decomposition.

```python
import numpy as np
n=200; t=np.arange(n)
trend=0.05*t; seasonal=3*np.sin(2*np.pi*t/20); noise=np.random.normal(0,0.5,n)
ts=trend+seasonal+noise; print('Shape:',ts.shape,'  Mean:',ts.mean().round(2))
```

```python
import matplotlib.pyplot as plt
trend_est=np.convolve(ts,np.ones(21)/21,mode='same')
fig,axes=plt.subplots(3,1,figsize=(10,6),sharex=True)
axes[0].plot(ts,label='Series'); axes[0].plot(trend_est,color='red',label='Trend'); axes[0].legend()
axes[1].plot(ts-trend_est,label='Detrended'); axes[1].legend()
x=ts-ts.mean(); c0=np.dot(x,x)/n
acf=[np.dot(x[:n-k],x[k:])/(n*c0) for k in range(41)]
axes[2].bar(range(41),acf); axes[2].set_title('ACF')
plt.tight_layout(); plt.show()
```
