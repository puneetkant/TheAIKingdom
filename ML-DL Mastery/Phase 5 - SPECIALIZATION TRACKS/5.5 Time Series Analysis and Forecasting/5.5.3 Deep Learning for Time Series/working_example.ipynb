# 5.5.3 Deep Learning for Time Series

LSTM, Temporal CNN, Transformer, N-HiTS, PatchTST, sliding windows.

```python
import numpy as np
# Create sliding-window dataset
n=300; t=np.arange(n); ts=np.sin(2*np.pi*t/30)+0.1*t/n+np.random.normal(0,.2,n)
look_back=20; X=[ts[i:i+look_back] for i in range(n-look_back-1)]; y=[ts[i+look_back] for i in range(n-look_back-1)]
X,y=np.array(X),np.array(y); print('X shape:',X.shape,'y shape:',y.shape)
```

```python
from sklearn.neural_network import MLPRegressor; from sklearn.preprocessing import StandardScaler
sc=StandardScaler(); X_s=sc.fit_transform(X); y_s=(y-y.mean())/y.std()
mlp=MLPRegressor(hidden_layer_sizes=(32,16),max_iter=200,random_state=0).fit(X_s[:200],y_s[:200])
pred=mlp.predict(X_s[200:])*y.std()+y.mean()
import matplotlib.pyplot as plt
plt.plot(y[200:],label='True'); plt.plot(pred,label='MLP'); plt.legend(); plt.title('TS Forecast'); plt.show()
```
