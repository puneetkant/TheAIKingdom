# 5.3.2 Tabular Methods

Q-Learning, SARSA, Monte Carlo, exploration strategies.

```python
import numpy as np
N=4; N_STATES=N*N; N_ACTIONS=4; GAMMA=0.9; ALPHA=0.1; EPS=0.2
def step(s,a):
    if s==N_STATES-1: return s,0
    r,c=divmod(s,N)
    if a==0: r=max(0,r-1)
    elif a==1: r=min(N-1,r+1)
    elif a==2: c=max(0,c-1)
    else: c=min(N-1,c+1)
    ns=r*N+c; return ns,(1 if ns==N_STATES-1 else -0.01)
Q=np.zeros((N_STATES,N_ACTIONS)); print('Q table shape:',Q.shape)
```

```python
returns=[]
for ep in range(1000):
    s=0; total=0
    for _ in range(200):
        a=np.random.randint(N_ACTIONS) if np.random.rand()<EPS else Q[s].argmax()
        ns,r=step(s,a); Q[s,a]+=ALPHA*(r+GAMMA*Q[ns].max()-Q[s,a])
        s=ns; total+=r
        if ns==N_STATES-1: break
    returns.append(total)
import matplotlib.pyplot as plt
plt.plot(np.convolve(returns,np.ones(50)/50,'valid')); plt.title('Q-Learning Return'); plt.show()
```
