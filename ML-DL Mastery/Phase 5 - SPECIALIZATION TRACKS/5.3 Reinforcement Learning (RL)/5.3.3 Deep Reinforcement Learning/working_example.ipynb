# 5.3.3 Deep Reinforcement Learning

DQN, Policy Gradients (REINFORCE), Actor-Critic, PPO, replay buffers.

```python
import numpy as np
# Replay buffer
class ReplayBuffer:
    def __init__(self, capacity=1000):
        self.buf=[]; self.cap=capacity
    def push(self, t):
        if len(self.buf)>=self.cap: self.buf.pop(0)
        self.buf.append(t)
    def sample(self, n):
        idx=np.random.choice(len(self.buf),n,replace=False)
        return [self.buf[i] for i in idx]
buf=ReplayBuffer(); [buf.push((np.random.randn(4),np.random.randint(2),np.random.randn(),np.random.randn(4),False)) for _ in range(100)]
print('Buffer size:',len(buf)); print('Sample size:',len(buf.sample(16)))
```

```python
# REINFORCE policy gradient sketch
returns=[]; rewards=[0.1]*10; GAMMA=0.99
G=0
for r in reversed(rewards):
    G=r+GAMMA*G; returns.insert(0,G)
import matplotlib.pyplot as plt
plt.bar(range(len(returns)),returns); plt.title('Discounted Returns G_t'); plt.show()
```
