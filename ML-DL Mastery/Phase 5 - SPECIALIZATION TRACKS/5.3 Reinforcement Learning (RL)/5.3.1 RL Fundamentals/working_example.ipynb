# 5.3.1 RL Fundamentals

MDP, Bellman equations, policy evaluation, value iteration on GridWorld.

```python
import numpy as np
N=4; N_STATES=N*N; GAMMA=0.9
def step(s,a):
    if s in (0,N_STATES-1): return s,0
    r,c=divmod(s,N)
    if a==0: r=max(0,r-1)
    elif a==1: r=min(N-1,r+1)
    elif a==2: c=max(0,c-1)
    else: c=min(N-1,c+1)
    return r*N+c,-1
V=np.zeros(N_STATES); print('Initial V:',V.reshape(N,N))
```

```python
for i in range(200):
    delta=0
    for s in range(N_STATES):
        if s in (0,N_STATES-1): continue
        q=[step(s,a)[1]+GAMMA*V[step(s,a)[0]] for a in range(4)]
        new_v=max(q); delta=max(delta,abs(new_v-V[s])); V[s]=new_v
    if delta<1e-6: print(f'Converged iter {i+1}'); break
print('Value function:\n',V.reshape(N,N).round(2))
```
