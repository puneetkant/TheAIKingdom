```python
# Custom gym-compatible environment
import numpy as np
class SimpleMDP:
    def reset(self): self.s=0; return np.array([self.s],float)
    def step(self,a):
        self.s+=1 if a==1 else -1
        done=abs(self.s)>=5
        return np.array([self.s],float), (1 if done and self.s>0 else -1 if done else 0), done, {}
env=SimpleMDP(); s=env.reset(); print('Init state:',s)
for _ in range(10):
    s,r,done,_=env.step(1)
    print(f'  s={s[0]:+.0f} r={r}',end=' DONE' if done else ''); print()
    if done: break
```
