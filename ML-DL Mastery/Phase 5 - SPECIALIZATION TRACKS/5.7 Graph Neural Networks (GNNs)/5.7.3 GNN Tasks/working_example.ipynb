# 5.7.3 GNN Tasks

Node classification, link prediction, graph classification — three GNN paradigms.

```python
import numpy as np
# Mini GCN layer
def gcn(A, X, W):
    A_hat=A+np.eye(len(A)); d=A_hat.sum(1); D=np.diag(1/np.sqrt(d))
    return np.tanh((D@A_hat@D)@X@W)
A=np.array([[0,1,1,0],[1,0,1,0],[1,1,0,1],[0,0,1,0]],float)
X=np.random.randn(4,3); W=np.random.randn(3,2)*0.4; H=gcn(A,X,W)
print('Node embeddings H:'); print(H.round(3))
```

```python
# Link prediction via inner product
scores=H@H.T; print('Link scores (upper tri):')
for i in range(4):
    for j in range(i+1,4): print(f'  ({i},{j}): {scores[i,j]:.3f}')
```

```python
# Graph classification: sum readout
readout=H.sum(axis=0); pred=int(readout[0]>readout[1])
print('Graph-level readout:',readout.round(3),'→ class',pred)
```
