```python
import numpy as np
# Stochastic block model — community structure
def sbm(n=12, K=3, p_in=0.6, p_out=0.05, seed=1):
    rng=np.random.default_rng(seed); n_per=n//K
    labels=np.repeat(np.arange(K),n_per)
    A=np.zeros((n,n))
    for i in range(n):
        for j in range(i+1,n):
            p=p_in if labels[i]==labels[j] else p_out
            if rng.random()<p: A[i,j]=A[j,i]=1
    return A, labels
A,labels=sbm(); print('Edges:',int(A.sum()//2),'Labels:',labels)
```

```python
# Spectral embedding
D=np.diag(A.sum(1)); L=D-A
eigvals,eigvecs=np.linalg.eigh(L); Z=eigvecs[:,1:3]
import matplotlib.pyplot as plt
for k in range(3): plt.scatter(Z[labels==k,0],Z[labels==k,1],label=f'C{k}')
plt.legend(); plt.title('Spectral Embedding'); plt.show()
```
