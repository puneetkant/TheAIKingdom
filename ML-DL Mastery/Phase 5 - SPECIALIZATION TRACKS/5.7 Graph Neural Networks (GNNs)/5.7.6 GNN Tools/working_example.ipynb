# 5.7.6 GNN Tools

PyTorch Geometric, DGL, NetworkX — data objects, message passing, mini-batching.

```python
import numpy as np
# Minimal graph data object
class Data:
    def __init__(self, x, edges):
        self.x=np.array(x,float); self.edges=np.array(edges,int)
        self.N=len(x)
    def adj(self):
        A=np.zeros((self.N,self.N))
        A[self.edges[0],self.edges[1]]=1; return A
# Sample graph
edges=np.array([[0,1,2,1],[1,2,0,3]])
data=Data(np.random.randn(4,3),edges)
print('Nodes:',data.N,'Edges:',data.edges.shape[1],'\nAdj:\n',data.adj())
```

```python
# Message passing layer
def mp_layer(data, W):
    A=data.adj()+np.eye(data.N); D_inv=np.diag(1/A.sum(1))
    return np.tanh(D_inv@A@data.x@W)
W=np.random.randn(3,2)*0.4; H=mp_layer(data,W)
print('Output H:',H.round(3),'\nReadout (mean):',H.mean(axis=0).round(3))
```
