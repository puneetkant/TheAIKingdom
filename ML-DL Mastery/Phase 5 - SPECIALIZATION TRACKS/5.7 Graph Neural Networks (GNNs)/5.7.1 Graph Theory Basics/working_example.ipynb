# 5.7.1 Graph Theory Basics

Nodes, edges, adjacency matrix, Laplacian, BFS/DFS, degree, clustering coefficient.

```python
import numpy as np
A=np.array([[0,1,1,0],[1,0,1,1],[1,1,0,0],[0,1,0,0]],float)
degree=A.sum(axis=1); D=np.diag(degree); L=D-A
print('Adj:\n',A,'\nDegree:',degree,'\nLaplacian:\n',L)
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,3,figsize=(10,3))
axes[0].imshow(A,cmap='Blues'); axes[0].set_title('Adj')
axes[1].imshow(D,cmap='Oranges'); axes[1].set_title('Degree')
axes[2].imshow(L,cmap='RdBu'); axes[2].set_title('Laplacian')
for ax in axes: ax.set_xticks([]); ax.set_yticks([])
plt.tight_layout(); plt.show()
```
