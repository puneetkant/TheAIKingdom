```python
import numpy as np
# Epsilon-greedy exploration in RecSys
class EpsGreedyBandit:
    def __init__(self, n_items, eps=0.1):
        self.q=np.zeros(n_items); self.n=np.zeros(n_items); self.eps=eps
    def select(self):
        if np.random.rand()<self.eps: return np.random.randint(len(self.q))
        return self.q.argmax()
    def update(self, a, r):
        self.n[a]+=1; self.q[a]+=(r-self.q[a])/self.n[a]
bandit=EpsGreedyBandit(10); [bandit.update(bandit.select(),np.random.rand()) for _ in range(200)]
print('Q values:',bandit.q.round(2)); print('Best item:',bandit.q.argmax())
```

```python
# MMR diversity re-ranking
import numpy as np; from sklearn.metrics.pairwise import cosine_similarity
scores=np.array([1.0,0.9,0.85,0.8,0.3,0.2])
vecs=np.random.randn(6,8); sim=cosine_similarity(vecs)
selected=[]
for _ in range(4):
    if not selected:
        selected.append(np.argmax(scores))
    else:
        mmr=[0.5*scores[i]-0.5*max(sim[i,j] for j in selected) for i in range(6) if i not in selected]
        cands=[i for i in range(6) if i not in selected]; selected.append(cands[np.argmax(mmr)])
print('MMR top-4:',selected)
```
