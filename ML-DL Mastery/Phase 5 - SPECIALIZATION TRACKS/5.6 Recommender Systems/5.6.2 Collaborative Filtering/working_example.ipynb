# 5.6.2 Collaborative Filtering

User-user CF, item-item CF, matrix factorisation, ALS, SVD, implicit feedback.

```python
import numpy as np
rng=np.random.default_rng(0)
R=np.array([[5,3,0,1],[4,0,4,1],[1,1,0,5],[0,0,5,4],[0,1,5,4]],float)
R[R==0]=np.nan; print('Rating matrix:\n',R)
```

```python
# User-user cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
R_fill=np.nan_to_num(R); sim=cosine_similarity(R_fill)
print('User similarity matrix:\n',sim.round(2))
import matplotlib.pyplot as plt
plt.imshow(sim,cmap='Blues'); plt.colorbar(); plt.title('User Similarity'); plt.show()
```
