# 5.6.4 Hybrid Approaches

Weighted hybrid, switching, cascade, feature augmentation, LightFM, two-stage.

```python
import numpy as np
# Weighted hybrid: score = (1-alpha)*content + alpha*cf
content=np.array([1.0,0.8,0.3,0.2,0.1]); cf=np.array([0.4,0.6,0.9,0.7,0.2])
for a in [0.0,0.3,0.5,0.7,1.0]:
    hybrid=(1-a)*content+a*cf; print(f'alpha={a:.1f} hybrid={hybrid.round(2)}')
```

```python
import matplotlib.pyplot as plt
alphas=np.linspace(0,1,50); rmse_content=0.5+0.1*alphas; rmse_cf=0.3+0.2*(1-alphas)
rmse_hybrid=np.minimum(rmse_content,rmse_cf)*0.9  # hybrid is better
plt.plot(alphas,rmse_content,'--',label='Content only'); plt.plot(alphas,rmse_cf,'--',label='CF only')
plt.plot(alphas,rmse_hybrid,label='Hybrid'); plt.xlabel('alpha (CF weight)'); plt.ylabel('RMSE')
plt.title('Hybrid vs Single-signal RMSE'); plt.legend(); plt.show()
```
