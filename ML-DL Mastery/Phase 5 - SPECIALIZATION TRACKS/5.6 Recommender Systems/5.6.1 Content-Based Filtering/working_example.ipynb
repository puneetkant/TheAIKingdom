# 5.6.1 Content-Based Filtering

Item profiles, TF-IDF features, cosine similarity, user profiles.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity; import numpy as np
items=['action sci-fi hacker ai','action sci-fi robot ai','thriller sci-fi dreams','sci-fi space drama','action superhero']
titles=['Matrix','Terminator','Inception','Interstellar','Avengers']
vec=TfidfVectorizer().fit_transform(items); sim=cosine_similarity(vec)
print('Similarity to Matrix:',{titles[i]:round(sim[0,i],3) for i in range(len(titles))})
```

```python
import matplotlib.pyplot as plt
plt.imshow(sim,cmap='Blues'); plt.xticks(range(5),titles,rotation=30); plt.yticks(range(5),titles)
plt.colorbar(); plt.title('Content Similarity Matrix'); plt.tight_layout(); plt.show()
```
