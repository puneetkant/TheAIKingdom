# 5.6.3 Deep Learning Recommenders

NCF, two-tower models, session-based, BERT4Rec, embedding lookups.

```python
import numpy as np
# Embedding lookup
n_users,n_items,emb_dim=50,30,8
rng=np.random.default_rng(0)
U=rng.normal(0,.1,(n_users,emb_dim)); I=rng.normal(0,.1,(n_items,emb_dim))
u,i=3,7; score=np.dot(U[u],I[i])
print(f'Dot-product score for user {u}, item {i}: {score:.4f}')
```

```python
# Two-tower similarity matrix (users x items)
sim=U@I.T; print('Sim matrix shape:',sim.shape)
import matplotlib.pyplot as plt
plt.imshow(sim[:10,:15],cmap='RdBu',aspect='auto'); plt.colorbar()
plt.xlabel('Item'); plt.ylabel('User'); plt.title('Embedding Dot-Product Scores'); plt.show()
```
