# 5.6.6 Evaluation

Precision@k, Recall@k, NDCG, MRR, coverage, diversity, A/B testing.

```python
import numpy as np
def precision_at_k(rec,rel,k): return len(set(rec[:k])&set(rel))/k
def recall_at_k(rec,rel,k): return len(set(rec[:k])&set(rel))/(len(rel)+1e-8)
def ndcg_at_k(rec,rel,k):
    dcg=sum(1/np.log2(i+2) for i,r in enumerate(rec[:k]) if r in rel)
    idcg=sum(1/np.log2(i+2) for i in range(min(len(rel),k)))
    return dcg/(idcg+1e-8)
rec=[1,2,3,4,5]; rel={1,3,7}
print(f'P@3={precision_at_k(rec,rel,3):.3f} R@5={recall_at_k(rec,rel,5):.3f} NDCG@5={ndcg_at_k(rec,rel,5):.3f}')
```

```python
import matplotlib.pyplot as plt
ks=[1,3,5,10,20]; ps=[precision_at_k(list(range(20)),{1,3,7,12,18},k) for k in ks]
plt.plot(ks,ps,'o-'); plt.xlabel('k'); plt.ylabel('Precision@k')
plt.title('Precision@k curve'); plt.grid(True); plt.show()
```
