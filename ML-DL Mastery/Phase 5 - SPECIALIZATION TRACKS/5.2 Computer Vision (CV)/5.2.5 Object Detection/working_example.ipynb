# 5.2.5 Object Detection

IoU, NMS, anchor boxes, detection metrics (mAP).

```python
import numpy as np
def iou(a,b):
    xi1,yi1=max(a[0],b[0]),max(a[1],b[1])
    xi2,yi2=min(a[2],b[2]),min(a[3],b[3])
    inter=max(0,xi2-xi1)*max(0,yi2-yi1)
    return inter/((a[2]-a[0])*(a[3]-a[1])+(b[2]-b[0])*(b[3]-b[1])-inter+1e-9)
b1=[10,10,60,60]; b2=[12,12,62,62]; b3=[100,80,150,130]
print(f'IoU(b1,b2)={iou(b1,b2):.4f}  IoU(b1,b3)={iou(b1,b3):.4f}')
```

```python
def nms(boxes,scores,thresh=.5):
    order=np.argsort(scores)[::-1]; keep=[]
    while len(order):
        i=order[0]; keep.append(i)
        order=[j for j in order[1:] if iou(boxes[i],boxes[j])<thresh]
    return keep
boxes=np.array([b1,b2,b3]); scores=[.95,.85,.90]
print('Kept after NMS:',nms(boxes,scores))
```
