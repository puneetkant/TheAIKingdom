```python
# Numpy-based image pipeline demo
import numpy as np; import matplotlib.pyplot as plt; from sklearn.datasets import load_digits
digits=load_digits(); imgs=[digits.images[i]/16. for i in range(6)]
fig,ax=plt.subplots(1,6,figsize=(12,2))
for i,im in enumerate(imgs): ax[i].imshow(im,cmap='gray'); ax[i].axis('off')
plt.suptitle('Digits Dataset Sample'); plt.tight_layout(); plt.show()
```
