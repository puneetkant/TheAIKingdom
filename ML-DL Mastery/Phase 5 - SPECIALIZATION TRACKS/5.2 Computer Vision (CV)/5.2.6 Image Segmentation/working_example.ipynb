# 5.2.6 Image Segmentation

Semantic, instance, panoptic segmentation. K-means pixel clustering as baseline.

```python
import numpy as np; import matplotlib.pyplot as plt; from sklearn.cluster import KMeans
rng=np.random.default_rng(42)
img=np.zeros((32,64,3),dtype=np.float32)
img[:,:,2]=0.6+rng.random((32,64))*.1
img[5:20,5:25,0]=0.9; img[5:20,5:25,2]=0.1
img[10:28,35:55,1]=0.8; img[10:28,35:55,2]=0.2
pixels=img.reshape(-1,3); km=KMeans(n_clusters=3,n_init=5,random_state=42).fit(pixels)
seg=km.labels_.reshape(32,64)
print('Unique segments:',np.unique(seg))
```

```python
fig,ax=plt.subplots(1,2,figsize=(8,3))
ax[0].imshow(img); ax[0].set_title('Original'); ax[0].axis('off')
ax[1].imshow(seg,cmap='tab10'); ax[1].set_title('K-means Seg'); ax[1].axis('off')
plt.tight_layout(); plt.show()
```
