# 5.2.4 Image Classification

SVM on raw pixels vs PCA features. CNN architecture overview.

```python
from sklearn.datasets import load_digits; from sklearn.svm import SVC
from sklearn.model_selection import train_test_split; from sklearn.metrics import accuracy_score
digits=load_digits(); X=digits.data/16.; y=digits.target
Xt,Xe,yt,ye=train_test_split(X,y,test_size=.2,random_state=42)
svm=SVC(kernel='rbf',C=10).fit(Xt,yt); print('SVM accuracy:',accuracy_score(ye,svm.predict(Xe)).round(4))
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=32,random_state=42)
Xt2=pca.fit_transform(Xt); Xe2=pca.transform(Xe)
svm2=SVC(kernel='rbf',C=10).fit(Xt2,yt)
print('PCA(32)+SVM accuracy:',accuracy_score(ye,svm2.predict(Xe2)).round(4))
print('Explained variance:',pca.explained_variance_ratio_.sum().round(3))
```
