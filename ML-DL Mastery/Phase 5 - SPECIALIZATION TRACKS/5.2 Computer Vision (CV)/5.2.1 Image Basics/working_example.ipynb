# 5.2.1 Image Basics

Pixel values, channels, grayscale conversion, histograms, basic transformations.

```python
import numpy as np; import matplotlib.pyplot as plt
rng=np.random.default_rng(42)
img=np.zeros((64,64,3),dtype=np.float32)
img[:,:,0]=np.linspace(0,1,64)[None,:]
img[:,:,1]=np.linspace(0,1,64)[:,None]
img[:,:,2]=rng.random((64,64))*.3
print('Shape:',img.shape,'  Min:',img.min(),'  Max:',img.max())
```

```python
gray=0.299*img[:,:,0]+0.587*img[:,:,1]+0.114*img[:,:,2]
fig,ax=plt.subplots(1,2,figsize=(8,4))
ax[0].imshow(img); ax[0].set_title('RGB')
ax[1].imshow(gray,cmap='gray'); ax[1].set_title('Grayscale')
plt.tight_layout(); plt.show()
```
