# 5.2.8 3D Vision and Advanced Topics

Point clouds, depth estimation, NeRF, stereo vision, perspective projection.

```python
import numpy as np; import matplotlib.pyplot as plt; from mpl_toolkits.mplot3d import Axes3D
rng=np.random.default_rng(42)
theta=rng.uniform(0,2*np.pi,100); phi=rng.uniform(0,np.pi,100)
pts=np.stack([np.sin(phi)*np.cos(theta),np.sin(phi)*np.sin(theta),np.cos(phi)],axis=1)
fig=plt.figure(figsize=(5,4)); ax=fig.add_subplot(111,projection='3d')
ax.scatter(pts[:,0],pts[:,1],pts[:,2],s=5,alpha=.5); ax.set_title('Point Cloud (sphere)'); plt.show()
```

```python
# Perspective projection
pts_shifted=pts+np.array([0,0,3]); proj=pts_shifted[:,:2]/(pts_shifted[:,2:3])
plt.scatter(proj[:,0],proj[:,1],s=5,alpha=.5); plt.title('Perspective Projection'); plt.axis('equal'); plt.show()
```
