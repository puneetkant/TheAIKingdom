# 5.2.2 Classical Computer Vision

Sobel edges, Harris corners, HOG features, template matching.

```python
import numpy as np; from sklearn.datasets import load_digits
digits=load_digits(); img=digits.images[0]
Kx=np.array([[-1,0,1],[-2,0,2],[-1,0,1]],dtype=float); Ky=Kx.T
pad=np.pad(img,1)
def conv(K): return np.array([[(pad[i:i+3,j:j+3]*K).sum() for j in range(8)] for i in range(8)])
Gx=conv(Kx); Gy=conv(Ky); edges=np.sqrt(Gx**2+Gy**2)
print('Sobel max:',edges.max().round(2))
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,2,figsize=(6,3))
ax[0].imshow(img,cmap='gray'); ax[0].set_title('Original')
ax[1].imshow(edges,cmap='hot'); ax[1].set_title('Sobel Edges')
plt.tight_layout(); plt.show()
```
