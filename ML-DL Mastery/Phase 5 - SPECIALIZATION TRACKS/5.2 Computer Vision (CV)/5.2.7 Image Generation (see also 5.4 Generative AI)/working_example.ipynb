# 5.2.7 Image Generation

VAE, GAN, Diffusion models. Latent space sampling and generation pipelines.

```python
# VAE reparametrisation + sampling conceptual demo
import numpy as np; import matplotlib.pyplot as plt
rng=np.random.default_rng(42)
# Encoder: maps input to mu, log_var
mu=np.array([0.5,-0.3]); log_var=np.array([-1.0,-0.5])
eps=rng.standard_normal((200,2))
z=mu+eps*np.exp(0.5*log_var)  # reparametrised samples
plt.scatter(z[:,0],z[:,1],alpha=.3,label='z samples')
plt.scatter(*mu,c='red',s=100,zorder=5,label='mu')
plt.legend(); plt.title('VAE Latent Samples'); plt.axis('equal'); plt.show()
```

```python
# GAN concept: generator loss pushes D(G(z)) → 1
D_real=np.array([0.8,0.9,0.7]); D_fake=np.array([0.3,0.2,0.4])
d_loss=-(np.log(D_real+1e-9)+np.log(1-D_fake+1e-9)).mean()
g_loss=-np.log(D_fake+1e-9).mean()
print(f'Discriminator loss: {d_loss:.4f}  Generator loss: {g_loss:.4f}')
```
