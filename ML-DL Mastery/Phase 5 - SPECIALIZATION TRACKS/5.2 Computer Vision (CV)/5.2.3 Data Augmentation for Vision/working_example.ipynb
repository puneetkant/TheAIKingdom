# 5.2.3 Data Augmentation for Vision

Flips, rotations, noise, colour jitter, cutout — augmentation pipeline.

```python
import numpy as np; import matplotlib.pyplot as plt; from sklearn.datasets import load_digits
digits=load_digits(); img=digits.images[3]/16.0
augments={'Original':img,'H-Flip':img[:,::-1],'Rotate90':np.rot90(img),'Noise':np.clip(img+np.random.randn(*img.shape)*.05,0,1),'Bright':np.clip(img*1.8,0,1)}
fig,ax=plt.subplots(1,5,figsize=(12,3))
for i,(n,a) in enumerate(augments.items()): ax[i].imshow(a,cmap='gray'); ax[i].set_title(n); ax[i].axis('off')
plt.tight_layout(); plt.show()
```
