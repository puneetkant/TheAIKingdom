# 5.1.4 Subword Tokenization

BPE merge algorithm. Handles OOV and morphological variants.

```python
from collections import Counter, defaultdict; import re
corpus=['low lower lowest','newer new news','wider wide']
def get_vocab(c):
    v=Counter()
    for s in c:
        for w in s.split(): v[' '.join(w)+' </w>']+=1
    return v
vocab=get_vocab(corpus); print(dict(vocab))
```

```python
def get_stats(v):
    p=defaultdict(int)
    for w,f in v.items():
        sym=w.split()
        for i in range(len(sym)-1): p[(sym[i],sym[i+1])]+=f
    return p
def merge(pair,v):
    return {re.sub(re.escape(' '.join(pair)),''.join(pair),w):f for w,f in v.items()}
for _ in range(8):
    s=get_stats(vocab)
    if not s: break
    best=max(s,key=s.get); vocab=merge(best,vocab); print('Merge:',''.join(best))
```
