# 5.1.5 Classical NLP Models

Naive Bayes, Logistic Regression, SVM for text classification with TF-IDF features.

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
cats=['sci.space','rec.sport.hockey','talk.politics.guns']
tr=fetch_20newsgroups(subset='train',categories=cats,remove=('headers','footers','quotes'))
te=fetch_20newsgroups(subset='test', categories=cats,remove=('headers','footers','quotes'))
tv=TfidfVectorizer(max_features=5000,sublinear_tf=True); Xt=tv.fit_transform(tr.data); Xe=tv.transform(te.data)
print('Train:',Xt.shape,'  Test:',Xe.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB; from sklearn.metrics import accuracy_score
nb=MultinomialNB(alpha=0.1).fit(Xt,tr.target)
print('NB accuracy:',accuracy_score(te.target,nb.predict(Xe)).round(4))
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000).fit(Xt,tr.target)
print('LR accuracy:',accuracy_score(te.target,lr.predict(Xe)).round(4))
```
