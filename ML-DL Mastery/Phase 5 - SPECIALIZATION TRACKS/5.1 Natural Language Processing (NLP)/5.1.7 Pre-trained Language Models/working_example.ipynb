# 5.1.7 Pre-trained Language Models

BERT, GPT, T5 — fine-tuning patterns using HuggingFace Transformers.

```python
# HuggingFace pipeline demo (requires transformers + torch)
try:
    from transformers import pipeline
    clf=pipeline('sentiment-analysis',model='distilbert-base-uncased-finetuned-sst-2-english')
    for text in ['I love this!','This is terrible.','Not bad at all.']:
        r=clf(text)[0]; print(f'[{r["label"]}:{r["score"]:.2f}] {text}')
except ImportError:
    print('Install: pip install transformers torch')
```

```python
# BERT fine-tuning pattern (conceptual)
print('''Fine-tuning BERT:
1. Load pretrained: AutoModel.from_pretrained("bert-base-uncased")
2. Add classification head: nn.Linear(768, num_classes)
3. Freeze lower layers (optional): for p in model.bert.parameters(): p.requires_grad = False
4. Train with low LR (2e-5 to 5e-5)
5. Evaluate on validation set with early stopping
''')
```
