# 5.1.2 Text Representation

Bag of Words, TF-IDF, N-grams, cosine similarity.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
corpus=['the cat sat on the mat','the dog chased the cat','machine learning is great for NLP','NLP uses machine learning']
tv=TfidfVectorizer(max_features=20); X=tv.fit_transform(corpus).toarray()
print('Shape:',X.shape,'  Features:',tv.get_feature_names_out()[:8])
```

```python
# Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(X); print('Cosine sim matrix:\n',sim.round(2))
```
