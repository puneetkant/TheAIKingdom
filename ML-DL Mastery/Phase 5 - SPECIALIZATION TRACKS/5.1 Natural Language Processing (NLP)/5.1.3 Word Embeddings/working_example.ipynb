# 5.1.3 Word Embeddings

CBOW from scratch, cosine similarity, nearest neighbours, PCA visualisation.

```python
import numpy as np
corpus='the cat sat on the mat the dog chased the cat machine learning is great'.split()
vocab=sorted(set(corpus)); w2i={w:i for i,w in enumerate(vocab)}
V=len(vocab); dim=4
np.random.seed(42)
W=np.random.randn(V,dim)*0.01; print('Vocab:',vocab)
```

```python
# Co-occurrence matrix (GloVe-like)
window=2; cooc=np.zeros((V,V))
for i,w in enumerate(corpus):
    for j in corpus[max(0,i-window):i]+corpus[i+1:i+window+1]:
        cooc[w2i[w],w2i[j]]+=1
print('Co-occurrence matrix shape:',cooc.shape)
print('co-occ[cat]:',dict(zip(vocab,cooc[w2i['cat']].astype(int))))
```
