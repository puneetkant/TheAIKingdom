# 5.1.1 Text Preprocessing

Cleaning, tokenization, stopword removal, stemming pipeline.

```python
import re, string
from collections import Counter
corpus=['The cat sat on the mat!','NLP helps computers understand language.','Text preprocessing is important.']
clean=lambda t: re.sub(r'\s+',' ',re.sub(f'[{re.escape(string.punctuation)}]',' ',t.lower())).strip()
tokens=[clean(d).split() for d in corpus]
print(tokens)
```

```python
sw={'the','a','an','is','on','was','it','and','of','to'}
filtered=[[t for t in doc if t not in sw] for doc in tokens]
freq=Counter(w for d in filtered for w in d)
print('Most common:',freq.most_common(8))
```
