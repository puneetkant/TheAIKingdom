# 5.1.8 NLP Tasks

NER, POS tagging, text classification, question answering, summarization.

```python
import re
sentences=['Apple Inc. was founded by Steve Jobs in Cupertino.','Google announced its AI model in San Francisco.']
patterns={'ORG':[r'Apple Inc\.|Google'],'PER':[r'Steve Jobs'],'LOC':[r'Cupertino|San Francisco']}
for sent in sentences:
    print('\nText:',sent)
    for label,pats in patterns.items():
        for pat in pats:
            for m in re.finditer(pat,sent): print(f'  [{label}] {m.group()}')
```

```python
# Text classification benchmark
from sklearn.datasets import fetch_20newsgroups; from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression; from sklearn.metrics import accuracy_score
cats=['sci.space','rec.sport.hockey']; tr=fetch_20newsgroups(subset='train',categories=cats,remove=('headers','footers','quotes'))
te=fetch_20newsgroups(subset='test',categories=cats,remove=('headers','footers','quotes'))
v=TfidfVectorizer(max_features=5000); clf=LogisticRegression(max_iter=500).fit(v.fit_transform(tr.data),tr.target)
print('Accuracy:',accuracy_score(te.target,clf.predict(v.transform(te.data))).round(4))
```
