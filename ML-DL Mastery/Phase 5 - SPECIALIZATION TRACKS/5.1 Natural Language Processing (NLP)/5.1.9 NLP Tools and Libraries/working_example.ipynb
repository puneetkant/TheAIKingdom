```python
# Library detection
import importlib
libs=['nltk','spacy','sklearn','transformers','gensim']
for lib in libs:
    avail=importlib.util.find_spec(lib) is not None
    print(f'  {lib:15s}: {"✓ installed" if avail else "✗ not installed"}')
print('\nInstall missing: pip install nltk spacy scikit-learn transformers gensim')
```

```python
# sklearn NLP pipeline demo
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import fetch_20newsgroups
cats=['sci.space','rec.sport.hockey']
tr=fetch_20newsgroups(subset='train',categories=cats,remove=('headers','footers','quotes'))
te=fetch_20newsgroups(subset='test',categories=cats,remove=('headers','footers','quotes'))
pipe=Pipeline([('tfidf',TfidfVectorizer(max_features=5000)),('clf',LogisticRegression(max_iter=500))])
pipe.fit(tr.data,tr.target)
print('Pipeline accuracy:',pipe.score(te.data,te.target).round(4))
```
