# 5.4.4 Flow-Based Models and Flow Matching

Normalizing flows, change-of-variables, affine coupling, RealNVP, Glow, flow matching.

```python
import numpy as np
# Change of variables: log p(x) = log p(z) + log|det dz/dx|
def log_prob_gauss(z): return -0.5*np.sum(z**2,axis=-1) - z.shape[-1]*0.5*np.log(2*np.pi)
# Affine coupling forward
s=np.array([0.5,-0.3]); t=np.array([0.2,0.5])
x=np.random.randn(100,2); z=x*np.exp(s)+t; log_det=np.sum(s)
log_px=log_prob_gauss(z)+log_det; print('Mean log p(x):',log_px.mean().round(4))
```

```python
import matplotlib.pyplot as plt
# Flow matching: linear interpolation trajectory x0 -> x1
x0=np.random.randn(50,2); x1=np.random.randn(50,2)+np.array([3,3])
ts=np.linspace(0,1,6)
fig,axes=plt.subplots(1,6,figsize=(14,2.5))
for ax,t in zip(axes,ts):
    xt=(1-t)*x0+t*x1; ax.scatter(xt[:,0],xt[:,1],s=5); ax.set_title(f't={t:.1f}'); ax.set_xlim(-4,7); ax.set_ylim(-4,7)
plt.suptitle('Flow Matching: Linear Interpolation'); plt.tight_layout(); plt.show()
```
