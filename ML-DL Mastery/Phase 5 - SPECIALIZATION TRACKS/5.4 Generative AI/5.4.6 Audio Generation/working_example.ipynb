# 5.4.6 Audio Generation

WaveNet, SoundStream, VALL-E, MusicLM, mel-spectrogram, audio tokenization.

```python
import numpy as np
SR=16000; t=np.linspace(0,1,SR,endpoint=False)
# Generate A440 + harmonic overtones
freqs=[440,880,1320]; wave=sum(np.sin(2*np.pi*f*t)/(i+1) for i,f in enumerate(freqs))/len(freqs)
print(f'Waveform: {len(wave)} samples, range [{wave.min():.3f}, {wave.max():.3f}]')
```

```python
import matplotlib.pyplot as plt
# STFT magnitude
n_fft=512; hop=256; window=np.hanning(n_fft)
frames=[np.abs(np.fft.rfft(wave[i:i+n_fft]*window)) for i in range(0,len(wave)-n_fft,hop)]
spec=np.array(frames).T
fig,ax=plt.subplots(figsize=(10,3))
ax.imshow(np.log1p(spec[:80]),aspect='auto',origin='lower'); ax.set_title('Log Spectrogram')
plt.tight_layout(); plt.show()
```
