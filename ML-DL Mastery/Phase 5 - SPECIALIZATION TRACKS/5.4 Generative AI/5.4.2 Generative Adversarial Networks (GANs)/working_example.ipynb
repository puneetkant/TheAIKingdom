# 5.4.2 Generative Adversarial Networks (GANs)

Min-max game, discriminator, generator, mode collapse, WGAN, training instability.

```python
import numpy as np
# GAN loss formulation
def d_loss(d_real, d_fake):
    return -np.mean(np.log(d_real+1e-8) + np.log(1-d_fake+1e-8))
def g_loss(d_fake):
    return -np.mean(np.log(d_fake+1e-8))
d_real=np.array([0.9,0.8,0.85]); d_fake=np.array([0.1,0.2,0.15])
print('D loss:',round(d_loss(d_real,d_fake),4))
print('G loss:',round(g_loss(d_fake),4))
```

```python
import matplotlib.pyplot as plt
# Visualise D output at training start vs end
x=np.linspace(-2,5,200); d_untrained=np.full_like(x,0.5)
d_trained=1/(1+np.exp(-(x-2.5)))  # fake trained D
plt.plot(x,d_untrained,'--',label='Untrained D'); plt.plot(x,d_trained,label='Trained D')
plt.axvline(3,ls=':',color='g',label='Real data'); plt.legend()
plt.title('D(x) before/after training'); plt.xlabel('x'); plt.ylabel('D(x)'); plt.show()
```
