# 5.4.5 Autoregressive Models for Generation

GPT, PixelCNN, WaveNet, autoregressive factorisation, sampling strategies.

```python
from collections import Counter,defaultdict; import numpy as np
corpus='the quick brown fox jumps over the lazy dog '*20
bigram=defaultdict(Counter)
for i in range(len(corpus)-2):
    bigram[corpus[i:i+2]][corpus[i+2]]+=1
print('Bigram contexts:',len(bigram)); print("Top next chars after 'th':",bigram['th'].most_common(5))
```

```python
# Temperature sampling
def sample(ctx,T=1.0):
    ctr=bigram.get(ctx,Counter())
    if not ctr: return ' '
    chars,cnts=zip(*ctr.items()); logits=np.array(cnts,float)/T
    p=np.exp(logits-logits.max()); p/=p.sum()
    return np.random.choice(list(chars),p=p)
for T in [0.1,0.5,1.0,2.0]:
    text='th'
    for _ in range(50): text+=sample(text[-2:],T=T)
    print(f'  T={T}: {text}')
```
