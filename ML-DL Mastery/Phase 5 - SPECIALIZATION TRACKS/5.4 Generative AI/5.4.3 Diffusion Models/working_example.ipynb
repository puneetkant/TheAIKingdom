# 5.4.3 Diffusion Models

DDPM, DDIM, forward/reverse process, noise schedule, score matching.

```python
import numpy as np
T=200; betas=np.linspace(1e-4,0.02,T)
alphas=1-betas; abar=np.cumprod(alphas)
print(f'T={T}'); print(f'sqrt(abar_0)={np.sqrt(abar[0]):.4f} (almost 1 → signal)'); print(f'sqrt(abar_T)={np.sqrt(abar[-1]):.4f} (near 0 → noise)')
```

```python
import matplotlib.pyplot as plt
x0=np.random.randn(100,2)*0.5+np.array([2,2])   # 2D data
fig,axes=plt.subplots(1,5,figsize=(12,2.5))
for ax,t in zip(axes,[0,25,75,125,199]):
    ab=abar[t]; eps=np.random.randn(*x0.shape)
    xt=np.sqrt(ab)*x0+np.sqrt(1-ab)*eps
    ax.scatter(xt[:,0],xt[:,1],s=5,alpha=.5); ax.set_title(f't={t}'); ax.set_xlim(-4,6); ax.set_ylim(-4,6)
plt.suptitle('Forward Diffusion'); plt.tight_layout(); plt.show()
```
