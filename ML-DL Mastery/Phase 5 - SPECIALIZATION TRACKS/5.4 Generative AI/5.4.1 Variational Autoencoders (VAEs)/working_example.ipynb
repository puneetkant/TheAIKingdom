# 5.4.1 Variational Autoencoders (VAEs)

ELBO, reparametrisation trick, encoder-decoder, latent space, KL divergence.

```python
import numpy as np
# Reparametrisation trick
def reparametrize(mu, logvar):
    eps=np.random.randn(*mu.shape)
    return mu + eps * np.exp(0.5 * logvar)
mu=np.array([[0.5,-0.3]]); logvar=np.array([[0.2,-0.5]])
samples=np.vstack([reparametrize(mu,logvar) for _ in range(500)])
print('Mean of samples:',samples.mean(axis=0).round(3),'  Target:',mu[0])
```

```python
import matplotlib.pyplot as plt
# KL divergence: D_KL(N(mu,sigma^2) || N(0,1))
mus=np.linspace(-2,2,50); sig2=np.linspace(0.1,2,50); Mu,Sig=np.meshgrid(mus,sig2)
KL=0.5*(-np.log(Sig)+Sig+Mu**2-1)
plt.contourf(Mu,Sig,KL,20,cmap='RdYlGn_r'); plt.colorbar()
plt.xlabel('mu'); plt.ylabel('sigma^2'); plt.title('KL(N(mu,s2)||N(0,1))')
plt.show()
```
