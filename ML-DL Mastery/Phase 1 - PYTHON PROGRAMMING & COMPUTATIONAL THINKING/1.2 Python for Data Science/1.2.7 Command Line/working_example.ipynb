# 1.2.7 Command Line

Python CLI tooling: `argparse`, `subprocess`, `pathlib`, `os.environ`, pipeline runner.

```python
import subprocess, sys, os
from pathlib import Path

# System info
print('Python:', sys.version)
print('CWD:', Path.cwd())
print('PATH entries:', len(os.environ.get('PATH','').split(os.pathsep)))
r=subprocess.run(['git','--version'],capture_output=True,text=True)
print('Git:', r.stdout.strip() if r.returncode==0 else 'not found')
```

```python
# argparse basics
import argparse
parser=argparse.ArgumentParser(prog='train')
parser.add_argument('--lr', type=float, default=0.01)
parser.add_argument('--epochs', type=int, default=10)
parser.add_argument('--verbose', action='store_true')
# Simulate parsing (in notebook, args come from a list not sys.argv)
args=parser.parse_args(['--lr','0.001','--epochs','50','--verbose'])
print(f'lr={args.lr}, epochs={args.epochs}, verbose={args.verbose}')
```

```python
# subprocess — run system commands safely
def sh(cmd): return subprocess.run(cmd,capture_output=True,text=True).stdout.strip()
print(sh([sys.executable,'--version']))
print(sh([sys.executable,'-m','pip','list','--format=columns'])[:200])
```

```python
# pathlib file management
from pathlib import Path
tmp=Path('tmp_cli_demo'); tmp.mkdir(exist_ok=True)
(tmp/'file1.txt').write_text('hello')
(tmp/'file2.csv').write_text('a,b\n1,2\n')
for f in sorted(tmp.iterdir()):
    print(f.name, f.stat().st_size, 'bytes')
import shutil; shutil.rmtree(tmp); print('cleaned up')
```

```python
# os.environ + dotenv-style config
import os
os.environ['MODEL_LR']='0.005'
os.environ['MODEL_EPOCHS']='100'
def get_config(key, default=None):
    return os.environ.get(key, default)
print(f'LR={get_config("MODEL_LR")}, epochs={get_config("MODEL_EPOCHS")}')
# Clean up
del os.environ['MODEL_LR']; del os.environ['MODEL_EPOCHS']
```
