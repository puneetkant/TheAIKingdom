# 1.2.4 Exploratory Data Analysis (EDA)

Structured EDA pipeline on Titanic: data quality → univariate → bivariate → visualisations.

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, urllib.request
from pathlib import Path
DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'titanic.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/phihung/titanic/resolve/main/train.csv',dest)
df=pd.read_csv(dest)
print(df.shape); df.info()
```

```python
# Data quality
print('--- Null counts ---'); print(df.isnull().sum())
print('\n--- Duplicates ---', df.duplicated().sum())
print('\n--- Describe ---'); df.describe().round(2)
```

```python
# Univariate distributions
fig,axes=plt.subplots(1,3,figsize=(15,4))
df['Age'].hist(bins=30,ax=axes[0],color='steelblue'); axes[0].set_title('Age')
df['Fare'].clip(upper=200).hist(bins=30,ax=axes[1],color='coral'); axes[1].set_title('Fare (clipped @200)')
df['Pclass'].value_counts().sort_index().plot(kind='bar',ax=axes[2],color='seagreen'); axes[2].set_title('Pclass')
plt.suptitle('Univariate'); plt.tight_layout(); plt.show()
```

```python
# Bivariate: survival by groups
fig,axes=plt.subplots(1,3,figsize=(15,4))
df.groupby('Pclass')['Survived'].mean().plot(kind='bar',ax=axes[0]); axes[0].set_title('Survival by Class')
df.groupby('Sex')['Survived'].mean().plot(kind='bar',ax=axes[1],color=['coral','steelblue']); axes[1].set_title('Survival by Sex')
df['Age'].hist(bins=20,ax=axes[2],alpha=0.6,density=True,label='all')
df[df['Survived']==1]['Age'].hist(bins=20,ax=axes[2],alpha=0.5,density=True,label='survived')
axes[2].legend(); axes[2].set_title('Age: all vs survived')
plt.suptitle('Bivariate'); plt.tight_layout(); plt.show()
```

```python
# Correlation heatmap
numeric=df[['Survived','Pclass','Age','SibSp','Parch','Fare']].dropna()
corr=numeric.corr()
fig,ax=plt.subplots(figsize=(7,5))
im=ax.imshow(corr.values,cmap='coolwarm',vmin=-1,vmax=1)
plt.colorbar(im)
ax.set_xticks(range(len(corr))); ax.set_xticklabels(corr.columns,rotation=45)
ax.set_yticks(range(len(corr))); ax.set_yticklabels(corr.columns)
for i in range(len(corr)):
    for j in range(len(corr)):
        ax.text(j,i,f'{corr.values[i,j]:.2f}',ha='center',va='center',fontsize=8)
ax.set_title('Correlation Matrix'); plt.tight_layout(); plt.show()
```
