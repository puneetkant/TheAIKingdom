# 1.2.3 Data Visualization

Matplotlib charts for ML: distributions, correlation heatmap, learning curves, confusion matrix.

```python
import numpy as np, matplotlib.pyplot as plt, csv, urllib.request
from pathlib import Path

DATA=Path('data'); DATA.mkdir(exist_ok=True)
dest=DATA/'cal_housing.csv'
if not dest.exists():
    urllib.request.urlretrieve('https://huggingface.co/datasets/scikit-learn/california-housing/resolve/main/cal_housing.csv',dest)

cols=['MedInc','HouseAge','AveRooms','AveBedrms','Population','AveOccup']
with open(dest,newline='') as f: rows=list(csv.DictReader(f))
X=np.array([[float(r[c]) for c in cols] for r in rows])
y=np.array([float(r['MedHouseVal']) for r in rows])
print(f'X:{X.shape} y:{y.shape}')
```

```python
# Distribution histograms
fig,axes=plt.subplots(2,3,figsize=(14,7))
for ax,col,i in zip(axes.flatten(),cols,range(6)):
    ax.hist(X[:,i],bins=40,color='steelblue',alpha=0.8)
    ax.set_title(col)
plt.suptitle('California Housing — Feature Distributions'); plt.tight_layout(); plt.show()
```

```python
# Scatter: income vs house value, coloured by age
fig,ax=plt.subplots(figsize=(8,5))
sc=ax.scatter(X[:,0],y,c=X[:,1],cmap='plasma',alpha=0.3,s=4)
plt.colorbar(sc,ax=ax,label='House Age')
ax.set_xlabel('MedInc'); ax.set_ylabel('MedHouseVal')
ax.set_title('Income vs House Value (colour=Age)')
plt.tight_layout(); plt.show()
```

```python
# Correlation heatmap
corr=np.corrcoef(X.T)
fig,ax=plt.subplots(figsize=(7,6))
im=ax.imshow(corr,cmap='coolwarm',vmin=-1,vmax=1)
plt.colorbar(im)
ax.set_xticks(range(6)); ax.set_xticklabels(cols,rotation=45,ha='right')
ax.set_yticks(range(6)); ax.set_yticklabels(cols)
for i in range(6):
    for j in range(6):
        ax.text(j,i,f'{corr[i,j]:.2f}',ha='center',va='center',fontsize=8,
                color='white' if abs(corr[i,j])>0.5 else 'black')
ax.set_title('Correlation Matrix'); plt.tight_layout(); plt.show()
```

```python
# Learning curves
import math,random; random.seed(42)
ep=list(range(1,51))
tl=[math.exp(-e*.07)+random.gauss(0,.01) for e in ep]
vl=[math.exp(-e*.06)+random.gauss(0,.02)+.05 for e in ep]
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(ep,tl,label='Train Loss'); ax.plot(ep,vl,'--',label='Val Loss')
ax.set_xlabel('Epoch'); ax.set_ylabel('Loss')
ax.set_title('Learning Curves'); ax.legend(); ax.grid(alpha=.3)
plt.tight_layout(); plt.show()
```
