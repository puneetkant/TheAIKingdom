# 1.2.7 Command Line — Shell Basics

Unix shell concepts in Python: file search, grep/cut/head pipes, pushd/popd, task runner.

```python
import os, sys, fnmatch
from pathlib import Path

# pwd, ls equivalent
print('PWD:', Path.cwd())
for f in sorted(Path('.').iterdir())[:10]:
    marker = '/' if f.is_dir() else ''
    print(f'  {f.name}{marker}')
```

```python
# find *.py equivalent
base=Path('.')
py_files=sorted(base.rglob('*.py'))
print(f'Python files: {len(py_files)}')
for f in py_files[:5]: print(' ', f)
```

```python
# grep + cut + head as Python generators
def grep(lines,pattern):
    for l in lines:
        if pattern in l: yield l
def cut(lines,field,sep=','):
    for l in lines:
        parts=l.strip().split(sep)
        if field<len(parts): yield parts[field]
def head(lines,n=5):
    for i,l in enumerate(lines):
        if i>=n: break
        yield l

# Demo with inline data
data='alice,female,29\nbob,male,35\ncarol,female,27\ndave,male,41\n'
result=list(head(cut(grep(data.splitlines(keepends=True),'female'),1),3))
print('grep female | cut -f2 | head -3:', result)
```

```python
# pushd/popd equivalent
orig=Path.cwd()
tmp=Path('tmp_shell_demo'); tmp.mkdir(exist_ok=True)
os.chdir(tmp); print('In:', Path.cwd().name)
(Path('test.txt')).write_text('hello shell')
print('Files:', [f.name for f in Path('.').iterdir()])
os.chdir(orig); print('Back:', Path.cwd().name)
import shutil; shutil.rmtree(tmp)
```
