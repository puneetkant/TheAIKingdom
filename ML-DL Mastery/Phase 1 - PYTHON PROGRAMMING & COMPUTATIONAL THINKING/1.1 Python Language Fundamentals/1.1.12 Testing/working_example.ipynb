# 1.1.12 Testing

Unittest, subTest, mocking, and ML-specific behavioural tests.

```python
import math, unittest

def normalize(data):
    lo, hi = min(data), max(data)
    return [(x-lo)/(hi-lo or 1e-8) for x in data]

def accuracy(y_true, y_pred):
    if len(y_true)!=len(y_pred): raise ValueError('length mismatch')
    return sum(a==b for a,b in zip(y_true,y_pred))/len(y_true)

print('Functions defined')
```

```python
class TestNormalize(unittest.TestCase):
    def test_range(self):
        r = normalize([1,2,3,4,5])
        self.assertAlmostEqual(min(r), 0.0)
        self.assertAlmostEqual(max(r), 1.0)
    def test_length_preserved(self):
        self.assertEqual(len(normalize([1,2,3])), 3)
    def test_parametrized(self):
        for data in [[0,1],[−10,10],[100,200]]:
            with self.subTest(data=data):
                r = normalize(data)
                self.assertAlmostEqual(min(r), 0.0)
                self.assertAlmostEqual(max(r), 1.0)

loader = unittest.TestLoader()
suite  = loader.loadTestsFromTestCase(TestNormalize)
runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite)
```

```python
class TestAccuracy(unittest.TestCase):
    def test_perfect(self):
        self.assertEqual(accuracy([1,0,1],[1,0,1]), 1.0)
    def test_zero(self):
        self.assertEqual(accuracy([1,1],[0,0]), 0.0)
    def test_half(self):
        self.assertAlmostEqual(accuracy([1,0,1,0],[1,0,0,1]), 0.5)
    def test_mismatch_raises(self):
        with self.assertRaises(ValueError): accuracy([1,0],[1])

runner.run(unittest.TestLoader().loadTestsFromTestCase(TestAccuracy))
```

```python
# Mock external calls
from unittest.mock import patch, MagicMock
import urllib.request
with patch('urllib.request.urlretrieve') as mock_fn:
    mock_fn.return_value = ('/tmp/data.csv', MagicMock())
    urllib.request.urlretrieve('http://example.com/data.csv', '/tmp/data.csv')
    print('Called with:', mock_fn.call_args)
    mock_fn.assert_called_once()
```

```python
# Assert patterns
assert normalize([1,5]) == [0.0, 1.0], 'boundary values fail'
assert accuracy(['a','a'],['a','b']) == 0.5
print('All assertions passed')
```
