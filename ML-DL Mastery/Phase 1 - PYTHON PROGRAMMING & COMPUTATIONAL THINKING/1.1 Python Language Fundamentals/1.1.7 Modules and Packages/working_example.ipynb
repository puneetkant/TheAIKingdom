# 1.1.7 Modules and Packages

Build a mini `mlutils` package, import it, introspect it, and explore the standard library.

```python
import sys, os, importlib, inspect
from pathlib import Path

BASE = Path().resolve()
pkg  = BASE / 'mlutils'
pkg.mkdir(exist_ok=True)

(pkg/'__init__.py').write_text('from .metrics import accuracy, f1\n__version__ = "0.1.0"\n')
(pkg/'metrics.py').write_text(
    'def accuracy(y_true, y_pred):\n'
    '    return sum(a==b for a,b in zip(y_true,y_pred))/len(y_true)\n'
    'def f1(y_true, y_pred):\n'
    '    tp=sum(1 for a,b in zip(y_true,y_pred) if a==1==b)\n'
    '    fp=sum(1 for a,b in zip(y_true,y_pred) if b==1!=a)\n'
    '    fn=sum(1 for a,b in zip(y_true,y_pred) if a==1!=b)\n'
    '    p=tp/(tp+fp+1e-9); r=tp/(tp+fn+1e-9)\n'
    '    return 2*p*r/(p+r+1e-9)\n'
)
print([f.name for f in sorted(pkg.iterdir())])
```

```python
if str(BASE) not in sys.path: sys.path.insert(0, str(BASE))
import mlutils
print('version:', mlutils.__version__)
y_true = [1,0,1,1,0,1]
y_pred = [1,0,0,1,1,1]
print('accuracy:', round(mlutils.accuracy(y_true, y_pred), 4))
print('f1 score:', round(mlutils.f1(y_true, y_pred), 4))
```

```python
# Introspect the metrics module
import mlutils.metrics as m
for name, fn in inspect.getmembers(m, inspect.isfunction):
    print(f'{name}{inspect.signature(fn)}')
```

```python
# Dynamic import
mod = importlib.import_module('mlutils.metrics')
print(dir(mod))
```
