# 1.1.13 Algorithms & Complexity

Sorting, binary search, heap top-K, edit distance (DP) — with Big-O timing comparisons.

```python
import random, time, heapq, math
random.seed(0)

def timed(label, fn, *args):
    t0 = time.perf_counter()
    r  = fn(*args)
    print(f'{label:<35} {(time.perf_counter()-t0)*1000:>8.3f} ms')
    return r

n = 10_000
data = [random.random() for _ in range(n)]
print(f'n={n:,}')
timed('python sorted [Timsort]',   sorted, data)

def merge_sort(a):
    if len(a)<=1: return a
    m=len(a)//2
    def merge(l,r):
        res=[]; i=j=0
        while i<len(l) and j<len(r):
            if l[i]<=r[j]: res.append(l[i]); i+=1
            else: res.append(r[j]); j+=1
        return res+l[i:]+r[j:]
    return merge(merge_sort(a[:m]),merge_sort(a[m:]))

timed('merge_sort [O(nlogn)]', merge_sort, data)
```

```python
# Binary search for lr
def find_lr(target, lo=1e-5, hi=1.0, tol=1e-7):
    loss = lambda lr: 1.0 - math.sqrt(lr)*0.9
    for _ in range(100):
        mid=( lo+hi)/2; v=loss(mid)
        if abs(v-target)<tol: return mid
        if v<target: hi=mid
        else: lo=mid
    return (lo+hi)/2

lr = find_lr(0.5)
print(f'Target loss=0.5 → lr={lr:.8f}, achieved={1-0.9*math.sqrt(lr):.8f}')
```

```python
# Heap top-K vs sort
n=100_000; k=10; vals=[random.random() for _ in range(n)]
timed(f'heapq.nlargest({k}) [O(nlogk)]', heapq.nlargest, k, vals)
timed(f'sorted()[:k]        [O(nlogn)]', lambda v: sorted(v,reverse=True)[:k], vals)
```

```python
# Edit distance DP
def edit_dist(s1, s2):
    m,n=len(s1),len(s2)
    dp=[[0]*(n+1) for _ in range(m+1)]
    for i in range(m+1): dp[i][0]=i
    for j in range(n+1): dp[0][j]=j
    for i in range(1,m+1):
        for j in range(1,n+1):
            dp[i][j]=dp[i-1][j-1] if s1[i-1]==s2[j-1] else 1+min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])
    return dp[m][n]

for s1,s2 in [('kitten','sitting'),('transformer','transducer'),('','abc')]:
    print(f'edit_dist({s1!r:15},{s2!r:15}) = {edit_dist(s1,s2)}')
```

```python
# Big-O cheat sheet
print('Big-O Complexity Reference:')
for complexity, example in [
    ('O(1)',      'dict lookup, list index'),
    ('O(log n)',  'binary search'),
    ('O(n)',      'linear scan, sum()'),
    ('O(n log n)','merge sort, Timsort'),
    ('O(n²)',     'insertion sort, naive string search'),
    ('O(n·m)',    'edit distance DP'),
    ('O(2ⁿ)',     'naive subset enumeration'),
]: print(f'  {complexity:<12} {example}')
```
