# 3.1.1 ML Core Concepts

Bias-variance tradeoff, train/val/test splits, learning curves with California Housing.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
housing=fetch_california_housing(); X,y=housing.data,housing.target
X_tmp,X_test,y_tmp,y_test=train_test_split(X,y,test_size=.15,random_state=42)
X_train,X_val,y_train,y_val=train_test_split(X_tmp,y_tmp,test_size=.15/.85,random_state=42)
print(f'Train:{X_train.shape[0]} Val:{X_val.shape[0]} Test:{X_test.shape[0]}')
```

```python
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
import numpy as np
scaler=StandardScaler().fit(X_train)
Xf_tr=scaler.transform(X_train)[:,:1]; Xf_te=scaler.transform(X_test)[:,:1]
for d in [1,2,3,5,9]:
    p=make_pipeline(PolynomialFeatures(d),Ridge(1e-6)); p.fit(Xf_tr,y_train)
    print(f'deg={d}: train={mean_squared_error(y_train,p.predict(Xf_tr)):.4f} test={mean_squared_error(y_test,p.predict(Xf_te)):.4f}')
```

```python
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
pipe=make_pipeline(StandardScaler(),Ridge(1.))
sizes,tr,va=learning_curve(pipe,X_train,y_train,cv=5,scoring='neg_mean_squared_error',train_sizes=np.linspace(.1,1.,8))
plt.plot(sizes,-tr.mean(1),label='Train'); plt.plot(sizes,-va.mean(1),label='Val')
plt.xlabel('n'); plt.ylabel('MSE'); plt.legend(); plt.title('Learning curves'); plt.show()
```

```python
# Final test evaluation
pipe=make_pipeline(StandardScaler(),Ridge(1.)); pipe.fit(X_train,y_train)
test_mse=mean_squared_error(y_test,pipe.predict(X_test))
print(f'Test MSE: {test_mse:.4f}  RMSE: {test_mse**.5:.4f}')
```
