# 3.5.4 Clustering Metrics

Silhouette Score, Davies-Bouldin Index, Calinski-Harabasz — internal metrics without ground truth.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score,davies_bouldin_score,calinski_harabasz_score
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
print(f'{"k":>4} {"Silhouette":>12} {"DBI":>10} {"CHI":>12}')
for k in range(2,9):
    lb=KMeans(k,random_state=42,n_init=5).fit_predict(X)
    print(f'{k:4d} {silhouette_score(X,lb,sample_size=2000,random_state=42):12.4f} {davies_bouldin_score(X,lb):10.4f} {calinski_harabasz_score(X,lb):12.1f}')
```

```python
from sklearn.metrics import adjusted_rand_score,normalized_mutual_info_score
import pandas as pd
df=pd.DataFrame(h.data,columns=h.feature_names)
true_l=((df.Latitude>df.Latitude.median()).astype(int)*2+(df.Longitude>df.Longitude.median()).astype(int))
pred_l=KMeans(4,random_state=42,n_init=10).fit_predict(X)
print(f'ARI={adjusted_rand_score(true_l,pred_l):.4f}  NMI={normalized_mutual_info_score(true_l,pred_l):.4f}')
```
