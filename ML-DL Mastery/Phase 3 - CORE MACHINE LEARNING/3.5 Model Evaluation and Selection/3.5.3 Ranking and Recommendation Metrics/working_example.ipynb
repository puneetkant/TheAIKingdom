# 3.5.3 Ranking & Recommendation Metrics

NDCG, MAP, Hit Rate — manual implementations and regression-as-ranking evaluation.

```python
import numpy as np
def dcg(r,k=None):
    r=np.asarray(r[:k],float); return np.sum(r/np.log2(np.arange(2,len(r)+2))) if len(r) else 0.
def ndcg(r,ideal,k=None):
    id2=sorted(ideal,reverse=True); d=dcg(id2,k); return dcg(r,k)/d if d else 0.
def ap(r): hits=ap_=0.;[exec('hits+=1;ap_+=hits/(i+1)') for i,v in enumerate(r) if v]; return ap_/max(1,hits) if hits else 0.

a=[1,1,0,1,0,1,0,0,1,0]; b=[0,0,1,0,1,0,1,1,0,1]
for name,r in [('A',a),('B',b)]:
    for k in [3,5,10]: print(f'{name} NDCG@{k}={ndcg(r,a,k):.4f}  AP@{k}={ap(r[:k]):.4f}')
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import make_pipeline
h=fetch_california_housing(); X,y=h.data,h.target
Xt,Xv,yt,yv=train_test_split(X,y,test_size=.2,random_state=42)
p=make_pipeline(StandardScaler(),GradientBoostingRegressor(100,random_state=42)); p.fit(Xt,yt)
yp=p.predict(Xv); rank=np.argsort(-yp); top20=set(np.where(yv>=np.percentile(yv,80))[0])
for k in [20,50,100]: print(f'Recall top-20% in top-{k}: {len(top20&set(rank[:k]))/len(top20):.2%}')
```
