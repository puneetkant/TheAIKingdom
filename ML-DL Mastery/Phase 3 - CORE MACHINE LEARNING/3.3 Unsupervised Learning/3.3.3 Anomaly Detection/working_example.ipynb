# 3.3.3 Anomaly Detection

Isolation Forest, Local Outlier Factor, One-Class SVM on California Housing.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
y_true=np.zeros(len(h.target),dtype=int)
y_true[(h.target>=np.percentile(h.target,95))|(h.target<=np.percentile(h.target,5))]=1
print(f'Anomalies: {y_true.sum()}/{len(y_true)}')
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import precision_score,recall_score
for c in [.05,.1,.15]:
    iso=IsolationForest(contamination=c,random_state=42,n_jobs=-1)
    a=(iso.fit_predict(X)==-1).astype(int)
    print(f'c={c}: prec={precision_score(y_true,a,zero_division=0):.4f} rec={recall_score(y_true,a,zero_division=0):.4f}')
```

```python
from sklearn.neighbors import LocalOutlierFactor
for k in [10,20,50]:
    lof=LocalOutlierFactor(k,contamination=.1)
    a=(lof.fit_predict(X)==-1).astype(int)
    print(f'k={k}: prec={precision_score(y_true,a,zero_division=0):.4f} rec={recall_score(y_true,a,zero_division=0):.4f}')
```

```python
from sklearn.svm import OneClassSVM
oc=OneClassSVM(nu=.05,kernel='rbf',gamma='scale')
a=(oc.fit_predict(X[:5000])==-1).astype(int)
print(f'OC-SVM prec={precision_score(y_true[:5000],a,zero_division=0):.4f} rec={recall_score(y_true[:5000],a,zero_division=0):.4f}')
```
