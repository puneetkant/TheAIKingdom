# 3.3.2 Dimensionality Reduction

PCA scree plot, cumulative variance, reconstruction error, PCA + Ridge pipeline.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
pca=PCA().fit(X); evr=pca.explained_variance_ratio_; cum=evr.cumsum()
for i,(e,c) in enumerate(zip(evr,cum),1): print(f'PC{i}: var={e:.4f} cum={c:.4f}')
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,2,figsize=(10,4))
axes[0].bar(range(1,len(evr)+1),evr); axes[0].set_title('Scree plot')
axes[1].plot(range(1,len(cum)+1),cum,'o-'); axes[1].axhline(.95,color='r',ls='--')
axes[1].set_title('Cumulative variance'); plt.show()
```

```python
# Reconstruction error
for k in [1,2,3,5,8]:
    p=PCA(k); Xr=p.inverse_transform(p.fit_transform(X))
    print(f'k={k}: recon_MSE={np.mean((X-Xr)**2):.4f} var={p.explained_variance_ratio_.sum():.4f}')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
Xr,yr=h.data,h.target
Xt,Xv,yt,yv=train_test_split(Xr,yr,test_size=.2,random_state=42)
for k in [2,4,6,8]:
    p=make_pipeline(StandardScaler(),PCA(k),Ridge(1.))
    p.fit(Xt,yt); print(f'PCA({k})+Ridge RMSE={mean_squared_error(yv,p.predict(Xv))**.5:.4f}')
```
