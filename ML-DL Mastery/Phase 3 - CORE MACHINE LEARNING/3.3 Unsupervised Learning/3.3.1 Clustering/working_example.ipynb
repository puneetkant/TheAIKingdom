# 3.3.1 Clustering

K-Means elbow/silhouette, DBSCAN, Agglomerative Clustering on California Housing.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
for k in range(2,8):
    km=KMeans(k,random_state=42,n_init=10); lbl=km.fit_predict(X)
    sil=silhouette_score(X,lbl,sample_size=2000,random_state=42)
    print(f'k={k}: inertia={km.inertia_:.1f} sil={sil:.4f}')
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
km=KMeans(4,random_state=42,n_init=10); lbl=km.fit_predict(X)
X2d=PCA(2).fit_transform(X)
plt.scatter(X2d[:,0],X2d[:,1],c=lbl,s=2,cmap='tab10',alpha=.3)
plt.title('K-Means k=4 (PCA projection)'); plt.colorbar(); plt.show()
```

```python
from sklearn.cluster import DBSCAN
Xs=X[:3000]
for eps in [.5,1.,1.5]:
    lbl=DBSCAN(eps=eps,min_samples=10).fit_predict(Xs)
    n=len(set(lbl))-(1 if -1 in lbl else 0)
    print(f'eps={eps}: clusters={n} noise={(lbl==-1).sum()}')
```

```python
from sklearn.cluster import AgglomerativeClustering
X2k=X[:2000]
for link in ['ward','average','complete']:
    lbl=AgglomerativeClustering(4,linkage=link).fit_predict(X2k)
    sil=silhouette_score(X2k,lbl,sample_size=1000,random_state=42)
    print(f'{link}: sil={sil:.4f}')
```
