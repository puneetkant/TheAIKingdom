# 3.2.3 K-Nearest Neighbors (KNN)

KNN regression k sweep, KNN classification AUC, curse of dimensionality illustration.

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import make_pipeline
import numpy as np
h=fetch_california_housing(); X,y=h.data,h.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
for k in [1,3,5,10,20,50]:
    p=make_pipeline(StandardScaler(),KNeighborsRegressor(k,weights='distance'))
    p.fit(X_train,y_train)
    print(f'k={k:>3}: RMSE={mean_squared_error(y_test,p.predict(X_test))**.5:.4f}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import roc_auc_score
y_bin=(y>np.median(y)).astype(int)
X_train2,X_test2,yb_train,yb_test=train_test_split(X,y_bin,test_size=.2,random_state=42)
for k in [3,5,10,20]:
    p=make_pipeline(StandardScaler(),KNeighborsClassifier(k))
    p.fit(X_train2,yb_train)
    print(f'k={k}: acc={p.score(X_test2,yb_test):.4f} AUC={roc_auc_score(yb_test,p.predict_proba(X_test2)[:,1]):.4f}')
```
