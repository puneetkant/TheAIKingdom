# 3.2.4 Naive Bayes

Gaussian NB on Cal Housing, Laplace smoothing illustration, text classification with MultinomialNB.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,classification_report
h=fetch_california_housing(); X,y=h.data,(h.target>np.median(h.target)).astype(int)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
gnb=GaussianNB().fit(X_train,y_train)
print('Accuracy:',accuracy_score(y_test,gnb.predict(X_test)).round(4))
```

```python
# Laplace smoothing
vocab=['cat','dog','fish']
X2=np.array([[1,1,0],[1,0,1],[0,1,1]]); y2=np.array([0,0,1])
for alpha in [0,.5,1.]:
    for cls in [0,1]:
        m=y2==cls; c=X2[m].sum(0)+alpha; t=X2[m].sum()+alpha*3
        print(f'α={alpha} cls={cls}: {dict(zip(vocab,(c/t).round(3)))}')
```

```python
# Text classification
try:
    from sklearn.datasets import fetch_20newsgroups
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.naive_bayes import MultinomialNB
    from sklearn.pipeline import make_pipeline
    cats=['sci.space','comp.graphics','rec.sport.baseball']
    tr=fetch_20newsgroups(subset='train',categories=cats,remove=('headers','footers','quotes'))
    te=fetch_20newsgroups(subset='test', categories=cats,remove=('headers','footers','quotes'))
    pipe=make_pipeline(TfidfVectorizer(max_features=5000),MultinomialNB())
    pipe.fit(tr.data,tr.target)
    print('Accuracy:',pipe.score(te.data,te.target).round(4))
except Exception as e:
    print('Dataset unavailable:',e)
```
