# 3.2.7 Ensemble Methods

Random Forest (OOB, importances), Gradient Boosting learning rate sweep, AdaBoost.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
h=fetch_california_housing(); X,y=h.data,h.target
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
for n in [10,50,100,200]:
    rf=RandomForestRegressor(n,oob_score=True,random_state=42,n_jobs=-1).fit(X_train,y_train)
    rmse=mean_squared_error(y_test,rf.predict(X_test))**.5
    oob=((y_train-rf.oob_prediction_)**2).mean()**.5
    print(f'n={n}: RMSE={rmse:.4f} OOB={oob:.4f}')
```

```python
import matplotlib.pyplot as plt
rf=RandomForestRegressor(200,random_state=42,n_jobs=-1).fit(X_train,y_train)
imps=rf.feature_importances_; names=h.feature_names; idx=imps.argsort()[::-1]
plt.bar(range(len(imps)),imps[idx])
plt.xticks(range(len(imps)),[names[i] for i in idx],rotation=45,ha='right')
plt.title('RF Feature Importances'); plt.tight_layout(); plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
for lr in [.01,.05,.1,.3]:
    gb=GradientBoostingRegressor(200,learning_rate=lr,max_depth=3,subsample=.8,random_state=42)
    gb.fit(X_train,y_train)
    print(f'lr={lr}: RMSE={mean_squared_error(y_test,gb.predict(X_test))**.5:.4f}')
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import roc_auc_score
y_bin=(h.target>np.median(h.target)).astype(int)
X2t,X2v,y2t,y2v=train_test_split(X,y_bin,test_size=.2,random_state=42)
ab=AdaBoostClassifier(100,random_state=42).fit(X2t,y2t)
print('AdaBoost AUC:',roc_auc_score(y2v,ab.predict_proba(X2v)[:,1]).round(4))
```
