# 3.2.5 Support Vector Machines (SVM)

Linear SVC with C sweep, RBF kernel SVC, SVR on California Housing.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
h=fetch_california_housing(); X,y=h.data[:5000],(h.target[:5000]>np.median(h.target[:5000])).astype(int)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=42)
for C in [.01,.1,1.,10.]:
    p=make_pipeline(StandardScaler(),LinearSVC(C=C,max_iter=5000))
    p.fit(X_train,y_train); print(f'C={C}: acc={p.score(X_test,y_test):.4f}')
```

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
pipe=make_pipeline(StandardScaler(),SVC(kernel='rbf',C=1.,gamma='scale',probability=True))
pipe.fit(X_train,y_train)
print(f'RBF SVC acc={pipe.score(X_test,y_test):.4f} AUC={roc_auc_score(y_test,pipe.predict_proba(X_test)[:,1]):.4f}')
```

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
h2=fetch_california_housing(); X2,y2=h2.data[:3000],h2.target[:3000]
Xt,Xv,yt,yv=train_test_split(X2,y2,test_size=.2,random_state=42)
for k in ['linear','rbf']:
    p=make_pipeline(StandardScaler(),SVR(kernel=k,C=1.,epsilon=.1))
    p.fit(Xt,yt); print(f'SVR({k}) RMSE={mean_squared_error(yv,p.predict(Xv))**.5:.4f}')
```
