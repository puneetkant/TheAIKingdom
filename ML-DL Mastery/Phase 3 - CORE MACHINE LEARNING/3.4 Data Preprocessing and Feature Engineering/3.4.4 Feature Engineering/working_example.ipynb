# 3.4.4 Feature Engineering

Polynomial features, log transforms, and domain-specific interaction features on California Housing.

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error,r2_score
h=fetch_california_housing(); X,y=h.data,h.target
Xt,Xv,yt,yv=train_test_split(X,y,test_size=.2,random_state=42)
p=make_pipeline(StandardScaler(),Ridge(1.)); p.fit(Xt,yt)
print('Baseline RMSE:',mean_squared_error(yv,p.predict(Xv))**.5:.4f)
```

```python
p2=make_pipeline(StandardScaler(),PolynomialFeatures(2,include_bias=False),StandardScaler(),Ridge(1.))
p2.fit(Xt,yt); n=p2.named_steps['polynomialfeatures'].n_output_features_
print(f'Poly(2) features: {n}, RMSE={mean_squared_error(yv,p2.predict(Xv))**.5:.4f}')
```

```python
# Log transform
from sklearn.preprocessing import FunctionTransformer
lp=make_pipeline(FunctionTransformer(lambda X:np.log1p(np.clip(X,0,None)),validate=True),StandardScaler(),Ridge(1.))
lp.fit(Xt,yt); print('Log1p RMSE:',mean_squared_error(yv,lp.predict(Xv))**.5:.4f)
```

```python
# Domain features
Xf=np.column_stack([X,X[:,2]/(X[:,5]+1),X[:,3]/(X[:,2]+1),X[:,4]/(X[:,5]+1)])
Xt2,Xv2,yt2,yv2=train_test_split(Xf,y,test_size=.2,random_state=42)
pd=make_pipeline(StandardScaler(),Ridge(1.)); pd.fit(Xt2,yt2)
print(f'Domain({Xf.shape[1]} feats) RMSE={mean_squared_error(yv2,pd.predict(Xv2))**.5:.4f}')
```
