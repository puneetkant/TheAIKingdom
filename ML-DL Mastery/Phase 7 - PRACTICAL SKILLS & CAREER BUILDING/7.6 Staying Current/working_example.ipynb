# 7.6 Staying Current

Simulates arXiv paper volume growth, a personalised reading schedule, and topic trend tracking.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)

years = np.arange(2015, 2025)
baseline = 500 * 2**((years-2015)/2)
paper_counts = baseline + rng.normal(0, baseline*0.08)

topics = {'Transformers':(2,0.50),'Diffusion':(0.1,0.90),'RL':(1,0.25),'LLMs':(0.2,0.85)}
t = np.arange(len(years))
trends = {name: s*np.exp(g*t)+rng.normal(0,s*0.1,len(t)) for name,(s,g) in topics.items()}

readable = 10/0.5  # papers per week at 10hr budget
backlog = max(0,(int(paper_counts[-1]/4)-readable)/int(paper_counts[-1]/4)*100)
print(f'Papers/month 2024 (sim): {paper_counts[-1]:,.0f}')
print(f'Readable/week: {readable:.0f}, Backlog rate: {backlog:.1f}%')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(years, paper_counts/1000, 'o-', color='steelblue', lw=2)
axes[0].fill_between(years, paper_counts/1000, alpha=0.15, color='steelblue')
axes[0].set(xlabel='Year', ylabel='Papers/Month (×1k)', title='arXiv Volume Growth')
axes[0].grid(True, alpha=0.3)

for name, vals in trends.items():
    axes[1].plot(years, vals/vals.max(), lw=2, label=name)
axes[1].set(xlabel='Year', ylabel='Normalised Popularity', title='AI Topic Trends')
axes[1].legend(fontsize=8); axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/staying_current.png')
print('Saved staying_current.png')
```
