# 7.3 Research Skills

Paper tracker, citation network analysis, and ablation study template.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)

papers = [
    {'title':'Attention Is All You Need','year':2017,'read':True,'citations':85000},
    {'title':'BERT','year':2018,'read':True,'citations':60000},
    {'title':'GPT-3','year':2020,'read':True,'citations':30000},
    {'title':'LoRA','year':2021,'read':True,'citations':8000},
    {'title':'Mamba','year':2023,'read':False,'citations':3000},
]

# Ablation study
baseline = 0.873
components = ['Attention', 'LayerNorm', 'Dropout', 'Pos Enc', 'FFN']
drops = [0.15, 0.05, 0.03, 0.08, 0.12]
ablation = {'Full': baseline}
for c, d in zip(components, drops): ablation[f'w/o {c}'] = baseline*(1-d)

for n, s in ablation.items(): print(f'{n:20s}: {s:.4f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 4))
axes[0].bar([p['title'][:12] for p in papers], [p['citations'] for p in papers],
            color=['mediumseagreen' if p['read'] else 'tomato' for p in papers])
axes[0].set(ylabel='Citations', title='Papers (green=read)'); axes[0].set_yscale('log'); axes[0].tick_params(axis='x',rotation=20)

names = list(ablation.keys()); scores = list(ablation.values())
colors = ['mediumseagreen'] + ['steelblue']*(len(names)-1)
axes[1].bar(names, scores, color=colors)
axes[1].set(ylabel='Accuracy', title='Ablation Study'); axes[1].set_ylim(0.7, 0.92)
axes[1].tick_params(axis='x', rotation=20); axes[1].grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('output/research_skills.png')
print('Saved research_skills.png')
```
