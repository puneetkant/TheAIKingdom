# 7.2 Portfolio Building

Score ML projects across five dimensions (documentation, code, tests, completeness, engagement) and visualise improvement impact.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)

projects = [
    {'name':'Image Classifier','readme':800,'files':5,'demo':1,'results':1,'stars':12,'commits':30,'tests':3},
    {'name':'NLP Sentiment','readme':1200,'files':8,'demo':1,'results':1,'stars':45,'commits':80,'tests':6},
    {'name':'RL Game Agent','readme':500,'files':3,'demo':0,'results':0,'stars':3,'commits':10,'tests':0},
    {'name':'Time Series','readme':950,'files':6,'demo':1,'results':1,'stars':22,'commits':45,'tests':4},
    {'name':'Kaggle Top-10','readme':1500,'files':12,'demo':1,'results':1,'stars':120,'commits':150,'tests':8},
]

def score(p):
    doc = min(p['readme']/1000,1)*25; code = min(p['files']/10,1)*20; test = min(p['tests']/5,1)*10
    comp = (p['demo']+p['results'])*10; eng = min(np.log1p(p['stars'])/5,1)*15+min(p['commits']/50,1)*10
    return doc+code+test+comp+eng

scores = [score(p) for p in projects]
for p, s in zip(projects, scores):
    print(f"{p['name']:20s}: {s:.1f}/100")
print(f'Portfolio mean: {np.mean(scores):.1f}/100')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
colors = plt.cm.RdYlGn(np.array(scores)/100)
axes[0].barh([p['name'] for p in projects], scores, color=colors)
axes[0].axvline(70, color='red', linestyle='--', alpha=0.5, label='Good (70)')
axes[0].set(xlabel='Score', title='Portfolio Project Scores'); axes[0].legend(); axes[0].grid(True, axis='x', alpha=0.3)

axes[1].scatter([p['stars'] for p in projects], scores, s=100, color='steelblue')
for p, s in zip(projects, scores):
    axes[1].annotate(p['name'][:8], (p['stars'], s), textcoords='offset points', xytext=(4,3), fontsize=7)
axes[1].set(xlabel='GitHub Stars', ylabel='Score', title='Stars vs Portfolio Score')
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/portfolio_building.png')
print('Saved portfolio_building.png')
```
