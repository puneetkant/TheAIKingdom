# 7.4 Interview Preparation

ML Q&A quiz scorer, time complexity table, and simulated practice performance over rounds.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)

QUESTIONS = [
    {'q':'Bias-variance tradeoff?', 'kw':['bias','variance','underfitting','overfitting'], 'diff':'Easy'},
    {'q':'Gradient descent variants?', 'kw':['sgd','adam','rmsprop','momentum'], 'diff':'Medium'},
    {'q':'Curse of dimensionality?', 'kw':['dimensionality','sparse','distance'], 'diff':'Medium'},
    {'q':'Attention mechanism?', 'kw':['query','key','value','softmax'], 'diff':'Medium'},
    {'q':'L1 vs L2 regularisation?', 'kw':['l1','lasso','l2','ridge','sparsity'], 'diff':'Easy'},
    {'q':'What is LoRA?', 'kw':['lora','rank','efficient','fine-tuning'], 'diff':'Hard'},
]

def score_ans(text, kw): return sum(k in text.lower() for k in kw)/len(kw)

# Simulate 5 practice rounds
rounds = []
for r in range(6):
    s = [np.clip(0.4+0.1*r+rng.normal(0,0.1),0,1) for _ in QUESTIONS]
    rounds.append(s)
means = [np.mean(r) for r in rounds]
for i, m in enumerate(means): print(f'Round {i+1}: mean score = {m:.3f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
final = rounds[-1]
axes[0].bar(range(1,len(QUESTIONS)+1), final,
            color=['mediumseagreen' if s>=0.7 else 'tomato' for s in final])
axes[0].axhline(0.7, color='k', linestyle='--', label='Target 0.7')
axes[0].set(xlabel='Question', ylabel='Score', title='Final Round Scores'); axes[0].legend(); axes[0].set_ylim(0,1.1)

axes[1].plot(range(1,7), means, 'o-', color='steelblue', lw=2)
axes[1].set(xlabel='Round', ylabel='Mean Score', title='Practice Progress'); axes[1].set_ylim(0,1); axes[1].grid(True,alpha=0.3)
plt.tight_layout()
plt.savefig('output/interview_prep.png')
print('Saved interview_prep.png')
```
