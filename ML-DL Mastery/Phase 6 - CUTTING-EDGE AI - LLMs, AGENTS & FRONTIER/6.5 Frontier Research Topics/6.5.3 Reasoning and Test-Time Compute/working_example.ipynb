# 6.5.3 Reasoning and Test-Time Compute

Simulate majority voting and best-of-N selection. Plot accuracy vs compute budget for different model strengths.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)

def sample_answer(true, p, rng):
    return true if rng.random()<p else rng.choice([a for a in 'ABCD' if a!=true])

def majority(answers):
    c = {};
    for a in answers: c[a] = c.get(a,0)+1
    return max(c, key=c.get)

def accuracy_vs_N(p, Ns, n_problems=100):
    true_ans = rng.choice(list('ABCD'), n_problems)
    accs = []
    for N in Ns:
        accs.append(sum(majority([sample_answer(t,p,rng) for _ in range(N)])==t for t in true_ans)/n_problems)
    return accs

Ns = [1,2,4,8,16,32]
for p in [0.4, 0.6, 0.8]:
    accs = accuracy_vs_N(p, Ns)
    print(f'p={p}: N=1→{accs[0]:.2f}, N=32→{accs[-1]:.2f}')
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
for p, color in zip([0.4, 0.6, 0.8], ['tomato','steelblue','mediumseagreen']):
    accs = accuracy_vs_N(p, Ns)
    ax.plot(Ns, accs, 'o-', color=color, lw=2, label=f'p={p}')
ax.set(xlabel='N (samples)', ylabel='Accuracy', title='Majority Vote: Accuracy vs Test-Time Compute')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/reasoning_compute.png')
print('Saved reasoning_compute.png')
```
