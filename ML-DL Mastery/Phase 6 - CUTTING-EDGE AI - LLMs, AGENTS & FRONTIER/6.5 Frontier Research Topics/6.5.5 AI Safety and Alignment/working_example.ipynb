# 6.5.5 AI Safety and Alignment

Simulate reward hacking (Goodhart's Law), KL divergence from a reference policy, and constitutional AI output filtering.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)

def simulate(n=300, beta=0.2):
    proxy,true_r,kl = [],[],[]
    for t in range(n):
        p = 0.5 + 0.5*(t/n) + rng.normal(0,.05)
        k = min(t/n*3, 3.) + rng.normal(0,.05)
        tr = 0.8*np.exp(-0.3*k) + rng.normal(0,.05)
        proxy.append(p); true_r.append(tr); kl.append(k)
    return np.array(proxy), np.array(true_r), np.array(kl)

proxy, true_r, kl = simulate()
print(f'Peak true reward at step: {true_r.argmax()}')
print(f'Final proxy: {proxy[-1]:.3f}, final true: {true_r[-1]:.3f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
steps = np.arange(len(proxy))
axes[0].plot(steps, proxy, color='tomato', lw=2, label='Proxy Reward')
axes[0].plot(steps, true_r, color='steelblue', lw=2, label='True Reward')
axes[0].axvline(true_r.argmax(), color='green', linestyle='--', label='True peak')
axes[0].set(xlabel='Step', ylabel='Reward', title="Reward Hacking (Goodhart's Law)")
axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(steps, kl, color='purple', lw=2)
axes[1].set(xlabel='Step', ylabel='KL(π||π_ref)', title='KL Divergence from Reference')
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/ai_safety.png')
print('Saved ai_safety.png')
```
