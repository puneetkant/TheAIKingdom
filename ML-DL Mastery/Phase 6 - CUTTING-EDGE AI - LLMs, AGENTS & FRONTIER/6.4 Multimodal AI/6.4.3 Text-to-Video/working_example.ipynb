# 6.4.3 Text-to-Video

Compute temporal consistency metrics between synthetic video frames and demonstrate linear frame interpolation.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)
base = rng.standard_normal((8, 8))

# Generate smooth video
frames = []
for t in range(16):
    f = np.roll(base, t//4, axis=1) + rng.normal(0, 0.1, (8, 8))
    frames.append(f)
frames = np.array(frames)

tc = [np.mean((frames[i+1]-frames[i])**2) for i in range(15)]
print(f'Video shape: {frames.shape}')
print(f'Mean temporal MSE: {np.mean(tc):.4f}')
```

```python
noisy = np.array([np.random.default_rng(i).standard_normal((8,8)) for i in range(16)])
tc_noisy = [np.mean((noisy[i+1]-noisy[i])**2) for i in range(15)]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(tc, 'o-', color='steelblue', lw=2, label='Smooth')
axes[0].plot(tc_noisy, 's--', color='tomato', lw=2, label='Random')
axes[0].set(xlabel='Frame Pair', ylabel='Temporal MSE', title='Temporal Consistency')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

sample = np.concatenate([frames[i] for i in [0,4,8,12]], axis=1)
axes[1].imshow(sample, cmap='gray', vmin=-2, vmax=2)
axes[1].set(title='Sample Frames t=0,4,8,12', xticks=[], yticks=[])
plt.tight_layout()
plt.savefig('output/text_to_video.png')
print('Saved text_to_video.png')
```
