# 6.1.3 Pre-training LLMs

Simulate next-token prediction with a character-level bigram model.
Track perplexity and cross-entropy loss as a function of training steps.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)
rng = np.random.default_rng(42)
vocab_size = 27
counts = rng.integers(1, 100, size=(vocab_size, vocab_size)).astype(float)
probs = counts / counts.sum(axis=1, keepdims=True)
print('Bigram prob matrix shape:', probs.shape)
print('Sample row sums:', probs.sum(axis=1)[:3].round(4))
```

```python
steps = np.arange(1, 201)
perplexity = 22 * np.exp(-0.025 * steps) + 5.5 + np.abs(rng.normal(0, 0.3, len(steps)))
loss = np.log(perplexity)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(steps, perplexity, color='steelblue')
axes[0].set(xlabel='Steps', ylabel='Perplexity', title='Perplexity vs Steps')
axes[0].grid(True, alpha=0.3)
axes[1].plot(steps, loss, color='tomato')
axes[1].set(xlabel='Steps', ylabel='Loss', title='Cross-Entropy Loss')
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/pretraining_curves.png')
print(f'Final perplexity: {perplexity[-1]:.2f}, Final loss: {loss[-1]:.4f}')
```

```python
fig2, ax = plt.subplots(figsize=(7, 6))
im = ax.imshow(probs[:12, :12], cmap='Blues')
ax.set(title='Bigram Transition Probabilities', xlabel='Next Token', ylabel='Current Token')
plt.colorbar(im, ax=ax)
plt.tight_layout()
plt.savefig('output/bigram_heatmap.png')
print('Saved bigram_heatmap.png')
```
