# 6.1.1 Transformer Architecture Deep Dive

Scaled dot-product attention, multi-head attention, positional encoding, layer norm.

```python
import numpy as np
def softmax(x): e=np.exp(x-x.max(-1,keepdims=True)); return e/e.sum(-1,keepdims=True)
def sdpa(Q,K,V):
    d=Q.shape[-1]; scores=Q@K.T/np.sqrt(d); return softmax(scores)@V
np.random.seed(0); seq,d=6,8
X=np.random.randn(seq,d); Wq=np.random.randn(d,d)*0.1; Wk=np.random.randn(d,d)*0.1; Wv=np.random.randn(d,d)*0.1
Q,K,V=X@Wq,X@Wk,X@Wv; out=sdpa(Q,K,V)
print('Q shape:',Q.shape,'Out shape:',out.shape)
```

```python
# Sinusoidal positional encoding
def pe(T,d):
    P=np.zeros((T,d)); pos=np.arange(T)[:,None]
    div=np.exp(np.arange(0,d,2)*(-np.log(10000)/d))
    P[:,0::2]=np.sin(pos*div); P[:,1::2]=np.cos(pos*div); return P
import matplotlib.pyplot as plt
plt.imshow(pe(32,16).T,aspect='auto',cmap='RdBu'); plt.title('Sinusoidal PE'); plt.colorbar(); plt.show()
```
