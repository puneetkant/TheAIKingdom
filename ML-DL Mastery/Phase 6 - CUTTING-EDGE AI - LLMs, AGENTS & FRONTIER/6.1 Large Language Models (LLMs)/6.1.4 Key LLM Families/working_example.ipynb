# 6.1.4 Key LLM Families

Compare parameter counts across major LLM families: GPT, LLaMA, Mistral, Gemma, Phi, Falcon, and more.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path
from matplotlib.patches import Patch

Path('output').mkdir(exist_ok=True)

models = ['GPT-3', 'GPT-4 est.', 'LLaMA-2 70B', 'LLaMA-3 70B',
          'Mistral 7B', 'Mixtral 8x7B', 'Gemma 7B', 'Phi-3', 'Falcon 40B']
params_b = [175, 1800, 70, 70, 7, 46.7, 7, 3.8, 40]
open_src = [False, False, True, True, True, True, True, True, True]

print('Models loaded:', len(models))
for m, p, o in zip(models, params_b, open_src):
    print(f'  {m:<20} {p:>8.1f}B  {"open" if o else "closed"}')
```

```python
colors = ['#2ecc71' if o else '#e74c3c' for o in open_src]
fig, ax = plt.subplots(figsize=(12, 5))
ax.bar(models, params_b, color=colors, edgecolor='white')
ax.set_yscale('log')
ax.set(ylabel='Parameters (B, log scale)', title='LLM Parameter Counts by Family')
ax.tick_params(axis='x', rotation=30)
legend = [Patch(color='#2ecc71', label='Open Source'), Patch(color='#e74c3c', label='Closed')]
ax.legend(handles=legend)
ax.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('output/llm_families.png')
print('Saved llm_families.png')
```
