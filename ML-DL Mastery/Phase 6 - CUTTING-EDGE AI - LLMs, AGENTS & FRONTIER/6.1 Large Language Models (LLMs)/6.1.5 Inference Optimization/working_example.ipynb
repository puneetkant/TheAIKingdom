# 6.1.5 Inference Optimization

Simulate KV-cache memory savings, batching throughput scaling, and greedy vs beam search decoding.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from pathlib import Path

Path('output').mkdir(exist_ok=True)

def kv_cache_mb(seq_len, n_heads=32, head_dim=128, n_layers=32, dtype_bytes=2):
    return 2 * seq_len * n_heads * head_dim * n_layers * dtype_bytes / (1024**2)

seq_lengths = np.arange(128, 8193, 128)
cache_mb = [kv_cache_mb(s) for s in seq_lengths]
print(f'KV-cache memory at 4096 tokens: {kv_cache_mb(4096):.1f} MB')
print(f'KV-cache memory at 8192 tokens: {kv_cache_mb(8192):.1f} MB')
```

```python
batch_sizes = [1, 2, 4, 8, 16, 32, 64]
throughput = [b * 150 / (1 + 0.02 * b) for b in batch_sizes]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(seq_lengths, cache_mb, color='steelblue', lw=2)
axes[0].set(xlabel='Sequence Length', ylabel='Memory (MB)', title='KV-Cache Memory')
axes[0].grid(True, alpha=0.3)

axes[1].plot(batch_sizes, throughput, 'o-', color='darkorange', lw=2)
axes[1].set(xlabel='Batch Size', ylabel='Tokens/sec', title='Batching Throughput')
axes[1].grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('output/inference_optimization.png')
print('Saved inference_optimization.png')
```
