# 6.1.2 Tokenisation for LLMs

BPE, WordPiece, SentencePiece — vocabulary building, encode/decode, special tokens.

```python
from collections import Counter
def get_pairs(vocab):
    pairs=Counter()
    for w,f in vocab.items():
        s=w.split()
        for i in range(len(s)-1): pairs[(s[i],s[i+1])]+=f
    return pairs
def merge(pair,vocab):
    bg=' '.join(pair); rep=''.join(pair)
    return {w.replace(bg,rep):f for w,f in vocab.items()}
# Mini corpus
vocab={'l o w </w>':5,'l o w e r </w>':2,'n e w e s t </w>':6,'w i d e s t </w>':3}
for _ in range(5):
    pairs=get_pairs(vocab)
    if not pairs: break
    best=pairs.most_common(1)[0][0]; vocab=merge(best,vocab)
    print('Merge:',best)
print('Final vocab:',list(vocab.keys()))
```

```python
# Fertility: avg tokens per word
words='the cat sat on the mat a lower lowest newest widest'.split()
fertility=[len(w)/3 for w in words]  # rough BPE estimate
import matplotlib.pyplot as plt
plt.bar(words,fertility); plt.xticks(rotation=45,ha='right'); plt.ylabel('Tokens'); plt.title('Estimated Token Fertility'); plt.tight_layout(); plt.show()
```
