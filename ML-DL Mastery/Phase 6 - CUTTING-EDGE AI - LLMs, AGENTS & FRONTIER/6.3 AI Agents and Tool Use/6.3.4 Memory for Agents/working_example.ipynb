# 6.3.4 Memory for Agents

Demonstrate three agent memory types: sliding window (in-context), episodic key-value store, and semantic search over embeddings.

```python
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
from collections import deque
from pathlib import Path

Path('output').mkdir(exist_ok=True)

# Sliding window memory
window = deque(maxlen=4)
msgs = ['Hello', 'Weather?', 'Sunny!', 'Great!', 'Let\'s go out.', 'Nice park.']
sizes = []
for m in msgs:
    window.append(m); sizes.append(len(window))
print('Final buffer:', list(window))

# Episodic memory
episodic = {'name': 'Alice', 'city': 'Paris', 'diet': 'vegetarian'}
for k in ['name', 'city', 'missing']:
    print(f'  Recall [{k}]: {episodic.get(k, "NOT FOUND")}')
```

```python
facts = [
    'The capital of France is Paris.',
    'Python is a programming language.',
    'Machine learning uses data to train models.',
    'Paris has the Eiffel Tower.',
    'Neural networks are inspired by the brain.',
]

def embed(text, dim=32):
    rng = np.random.default_rng(sum(ord(c) for c in text) % (2**31))
    v = rng.standard_normal(dim)
    return v / (np.linalg.norm(v) + 1e-10)

embs = np.array([embed(f) for f in facts])
query = 'Paris attractions'
qvec = embed(query)
scores = embs @ qvec
top = np.argsort(scores)[::-1][:3]
print(f'Semantic search: "{query}"')
for i in top:
    print(f'  [{scores[i]:.3f}] {facts[i]}')

fig, ax = plt.subplots(figsize=(9, 4))
colors = ['tomato' if i in top else 'steelblue' for i in range(len(facts))]
ax.barh(range(len(facts)), scores, color=colors)
ax.set(yticks=range(len(facts)), yticklabels=[f[:45] for f in facts],
       xlabel='Cosine Similarity', title=f'Semantic Memory Search: "{query}"')
ax.tick_params(axis='y', labelsize=8)
plt.tight_layout()
plt.savefig('output/memory_agents.png')
print('Saved memory_agents.png')
```
