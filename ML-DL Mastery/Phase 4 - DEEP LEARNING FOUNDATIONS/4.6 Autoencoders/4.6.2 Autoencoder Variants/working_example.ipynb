# 4.6.2 Autoencoder Variants

Denoising AE (noisy input → clean target), Sparse AE (L1 on latent), VAE (reparametrisation trick).

```python
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
relu=lambda x:np.maximum(0,x)
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
Xt,Xe=train_test_split(X,test_size=.2,random_state=42)
rng=np.random.default_rng(0); noise=.5
Xtn=Xt+rng.standard_normal(Xt.shape)*noise; Xen=Xe+rng.standard_normal(Xe.shape)*noise
print('Noisy train shape:',Xtn.shape)
```

```python
# VAE reparametrisation trick concept
rng2=np.random.default_rng(42); mu=rng2.standard_normal((5,3)); log_var=rng2.standard_normal((5,3))
eps=rng2.standard_normal(mu.shape)
z=mu+np.exp(.5*log_var)*eps  # reparametrize: z ~ N(mu, sigma)
kl_loss=-0.5*np.mean(1+log_var-mu**2-np.exp(log_var))  # KL(N(mu,sigma)||N(0,1))
print('Sampled z shape:',z.shape); print(f'KL loss: {kl_loss:.4f}')
```
