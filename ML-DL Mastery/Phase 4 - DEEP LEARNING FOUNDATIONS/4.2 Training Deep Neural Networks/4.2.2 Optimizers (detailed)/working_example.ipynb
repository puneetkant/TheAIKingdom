# 4.2.2 Optimizers (Detailed)

SGD, Momentum, RMSProp, Adam — numpy implementations, convergence comparison on f(x)=x².

```python
import numpy as np

class SGD:
    def __init__(self,lr): self.lr=lr
    def update(self,p,g): return p-self.lr*g

class Adam:
    def __init__(self,lr,b1=.9,b2=.999,eps=1e-8):
        self.lr,self.b1,self.b2,self.eps,self.m,self.v,self.t=lr,b1,b2,eps,0.,0.,0
    def update(self,p,g):
        self.t+=1; self.m=self.b1*self.m+(1-self.b1)*g; self.v=self.b2*self.v+(1-self.b2)*g**2
        mh=self.m/(1-self.b1**self.t); vh=self.v/(1-self.b2**self.t)
        return p-self.lr*mh/(np.sqrt(vh)+self.eps)

for name,opt in [('SGD',SGD(.1)),('Adam',Adam(.1))]:
    x=10.; xs=[x]
    for _ in range(50): x=opt.update(x,2*x); xs.append(x)
    print(f'{name}: final x={xs[-1]:.6f}')
```

```python
import matplotlib.pyplot as plt
for name,lr in [('SGD',.1),('Adam',.1)]:
    opt=SGD(lr) if name=='SGD' else Adam(lr)
    x=10.; xs=[x]
    for _ in range(50): x=opt.update(x,2*x); xs.append(x)
    plt.plot(xs,label=name)
plt.axhline(0,c='k',ls='--'); plt.xlabel('Iter'); plt.title('f(x)=x²'); plt.legend(); plt.show()
```
