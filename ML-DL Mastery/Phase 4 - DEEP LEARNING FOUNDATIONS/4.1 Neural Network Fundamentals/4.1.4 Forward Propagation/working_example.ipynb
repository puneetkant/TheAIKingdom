# 4.1.4 Forward Propagation

Matrix math for MLP forward pass: z = XW + b, activations, layer shapes.

```python
import numpy as np
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
sigmoid=lambda x:1/(1+np.exp(-np.clip(x,-500,500))); relu=lambda x:np.maximum(0,x)

class MLP:
    def __init__(self,sizes,seed=42):
        rng=np.random.default_rng(seed); self.params=[]
        for fin,fout in zip(sizes[:-1],sizes[1:]):
            self.params.append((rng.standard_normal((fin,fout))*np.sqrt(2/fin),np.zeros(fout)))
    def forward(self,X):
        a=X
        for i,(W,b) in enumerate(self.params):
            z=a@W+b; a=sigmoid(z) if i==len(self.params)-1 else relu(z)
            print(f'Layer {i+1}: z{z.shape} → a{a.shape}  mean={a.mean():.4f}')
        return a

X,y=make_moons(200,noise=.2,random_state=42); Xs=StandardScaler().fit_transform(X)
net=MLP([2,8,8,1]); out=net.forward(Xs[:4])
```

```python
print('Sample outputs:',out[:,0].round(4))
print('Predicted:     ',(out[:,0]>.5).astype(int).tolist())
print('True labels:   ',y[:4].tolist())
```

```python
print('=== Dimension walkthrough ===')
layers=[2,16,8,1]
for i,(fin,fout) in enumerate(zip(layers[:-1],layers[1:])): print(f'Layer {i+1}: W({fin},{fout})')
```
