# 4.1.2 Activation Functions

Sigmoid, Tanh, ReLU, Leaky ReLU, ELU, GELU, Softmax — values, gradients, visualisation.

```python
import numpy as np
sigmoid=lambda x:1/(1+np.exp(-x))
tanh_d=lambda x:1-np.tanh(x)**2
relu=lambda x:np.maximum(0,x)
gelu=lambda x:0.5*x*(1+np.tanh(np.sqrt(2/np.pi)*(x+0.044715*x**3)))
z=np.array([-2.,-1.,0.,1.,2.])
print(f'{"z":>6} {"sigmoid":>10} {"tanh":>10} {"relu":>10} {"gelu":>10}')
for zi in z: print(f'{zi:6.1f} {sigmoid(zi):10.4f} {np.tanh(zi):10.4f} {relu(zi):10.4f} {gelu(zi):10.4f}')
```

```python
import matplotlib.pyplot as plt
x=np.linspace(-4,4,300)
fns=[('Sigmoid',sigmoid(x),'b'),('Tanh',np.tanh(x),'g'),('ReLU',relu(x),'r'),('GELU',gelu(x),'m')]
fig,ax=plt.subplots(1,4,figsize=(14,3))
for a,(n,v,c) in zip(ax,fns):
    a.plot(x,v,c,lw=2); a.axhline(0,c='k',lw=.5); a.set_title(n); a.grid(alpha=.3)
plt.tight_layout(); plt.show()
```

```python
# Softmax demo
def softmax(z): e=np.exp(z-z.max()); return e/e.sum()
logits=np.array([2.,1.,.5]); probs=softmax(logits)
print(f'Logits: {logits}  →  Probs: {probs.round(4)}  sum={probs.sum():.6f}')
```
