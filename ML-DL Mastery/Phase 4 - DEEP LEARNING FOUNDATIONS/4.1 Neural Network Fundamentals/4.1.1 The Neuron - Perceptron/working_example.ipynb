# 4.1.1 The Neuron & Perceptron

McCulloch-Pitts neuron, Perceptron learning rule, linearly separable vs non-separable.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.preprocessing import StandardScaler

class Perceptron:
    def __init__(self,lr=.1,n_iter=100): self.lr,self.n_iter=lr,n_iter
    def fit(self,X,y):
        self.w=np.zeros(X.shape[1]); self.b=0.; self.errs=[]
        for _ in range(self.n_iter):
            e=0
            for xi,yi in zip(X,y):
                p=1 if np.dot(self.w,xi)+self.b>=0 else -1; d=self.lr*(yi-p)
                self.w+=d*xi; self.b+=d; e+=int(d!=0)
            self.errs.append(e)
    def predict(self,X): return np.array([1 if np.dot(self.w,x)+self.b>=0 else -1 for x in X])

X,y=make_classification(200,n_features=2,n_redundant=0,n_clusters_per_class=1,random_state=42)
y2=np.where(y==0,-1,1); Xs=StandardScaler().fit_transform(X)
p=Perceptron(); p.fit(Xs,y2); print('Accuracy:',(p.predict(Xs)==y2).mean())
```

```python
import matplotlib.pyplot as plt
plt.plot(p.errs); plt.xlabel('Epoch'); plt.ylabel('Misclassifications'); plt.title('Perceptron Training'); plt.show()
```

```python
from sklearn.datasets import make_circles
from sklearn.linear_model import Perceptron as SkP
Xc,yc=make_circles(300,noise=.1,random_state=42); Xcs=StandardScaler().fit_transform(Xc)
sk=SkP(max_iter=200,random_state=42); sk.fit(Xcs,yc)
print('sklearn Perceptron on circles (should fail):',sk.score(Xcs,yc))
```
