# 4.1.3 Loss Functions

MSE, Binary Cross-Entropy, Softmax CE, Hinge, Huber — manual implementations and visualisation.

```python
import numpy as np
mse=lambda y,yh:np.mean((y-yh)**2)
bce=lambda y,p,e=1e-7:-np.mean(y*np.log(np.clip(p,e,1-e))+(1-y)*np.log(np.clip(1-p,e,1-e)))
hinge=lambda y,s:np.maximum(0,1-y*s).mean()
huber=lambda y,yh,d=1:np.where(np.abs(y-yh)<=d,.5*(y-yh)**2,d*(np.abs(y-yh)-.5*d)).mean()
y=np.array([0.,1.,1.,0.,1.]); p=np.array([.1,.9,.4,.2,.8]); ypm=np.where(y==0,-1.,1.)
print(f'MSE={mse(y,p):.4f}  BCE={bce(y,p):.4f}  Hinge={hinge(ypm,p):.4f}  Huber={huber(y,p):.4f}')
```

```python
def softmax_ce(true_idx,logits):
    z=logits-logits.max(); return -(z-np.log(np.exp(z).sum()))[true_idx]
logits=np.array([2.,1.,.1])
for tc in range(3): print(f'True={tc}: CE={softmax_ce(tc,logits):.4f}')
```

```python
import matplotlib.pyplot as plt
p2=np.linspace(.01,.99,100); r=np.linspace(-3,3,200)
fig,ax=plt.subplots(1,2,figsize=(10,4))
ax[0].plot(p2,-np.log(p2),label='BCE(y=1)'); ax[0].plot(p2,-np.log(1-p2),'--',label='BCE(y=0)')
ax[0].set_ylim(0,5); ax[0].legend(); ax[0].set_title('BCE')
ax[1].plot(r,r**2/2,label='MSE'); ax[1].plot(r,np.abs(r),'--',label='MAE')
ax[1].set_ylim(0,4); ax[1].legend(); ax[1].set_title('Regression Losses')
plt.tight_layout(); plt.show()
```
