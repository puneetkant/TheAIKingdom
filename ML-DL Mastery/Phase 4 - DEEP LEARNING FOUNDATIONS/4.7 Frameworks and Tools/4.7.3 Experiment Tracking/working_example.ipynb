# 4.7.3 Experiment Tracking

Log params, metrics, artifacts. MLflow API concepts implemented with stdlib.

```python
import json, uuid, time, numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
Xt,Xe,yt,ye=train_test_split(X,h.target,test_size=.2,random_state=42)
```

```python
results=[]
for alpha in [0.001,0.01,0.1,1.0,10.0]:
    m=Ridge(alpha=alpha).fit(Xt,yt)
    r={'alpha':alpha,'train_mse':mean_squared_error(yt,m.predict(Xt)),'test_mse':mean_squared_error(ye,m.predict(Xe))}
    results.append(r); print(f'alpha={alpha:.3f}  test_mse={r["test_mse"]:.4f}')
best=min(results,key=lambda x:x['test_mse']); print(f'Best: {best}')
```
