# 4.7.2 TensorFlow / Keras

Sequential API, `model.fit()`, callbacks, evaluation.

```python
import tensorflow as tf; from tensorflow import keras
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
Xt,Xe,yt,ye=train_test_split(X,h.target,test_size=.2,random_state=42)
print('TF version:',tf.__version__)
```

```python
model=keras.Sequential([keras.layers.Dense(64,activation='relu',input_shape=(8,)),keras.layers.Dense(32,activation='relu'),keras.layers.Dense(1)])
model.compile(optimizer='adam',loss='mse',metrics=['mae'])
hist=model.fit(Xt,yt,validation_split=.1,epochs=50,batch_size=64,verbose=0)
print('Val MSE:',model.evaluate(Xe,ye,verbose=0))
```
