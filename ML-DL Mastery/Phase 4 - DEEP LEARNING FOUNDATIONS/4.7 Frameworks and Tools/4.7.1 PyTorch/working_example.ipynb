# 4.7.1 PyTorch

Tensors, autograd, nn.Module, training loop — the PyTorch workflow.

```python
import torch, torch.nn as nn, torch.optim as optim

# Autograd demo
x=torch.tensor([2.,3.],requires_grad=True)
y=(x**2).sum(); y.backward()
print('dy/dx:',x.grad.numpy(),'  (expected [4,6])')
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import numpy as np
h=fetch_california_housing(); X=StandardScaler().fit_transform(h.data)
Xt,Xe,yt,ye=train_test_split(X,h.target,test_size=.2,random_state=42)
Xt=torch.tensor(Xt,dtype=torch.float32); Xe=torch.tensor(Xe,dtype=torch.float32)
yt=torch.tensor(yt,dtype=torch.float32).unsqueeze(1); ye=torch.tensor(ye,dtype=torch.float32).unsqueeze(1)
```

```python
model=nn.Sequential(nn.Linear(8,64),nn.ReLU(),nn.Linear(64,32),nn.ReLU(),nn.Linear(32,1))
opt=optim.Adam(model.parameters(),lr=1e-3); loss_fn=nn.MSELoss(); losses=[]
for ep in range(150):
    model.train(); loss=loss_fn(model(Xt),yt); opt.zero_grad(); loss.backward(); opt.step(); losses.append(loss.item())
model.eval()
with torch.no_grad(): print(f'Val MSE: {loss_fn(model(Xe),ye).item():.4f}')
import matplotlib.pyplot as plt; plt.plot(losses); plt.title('PyTorch Training'); plt.show()
```
