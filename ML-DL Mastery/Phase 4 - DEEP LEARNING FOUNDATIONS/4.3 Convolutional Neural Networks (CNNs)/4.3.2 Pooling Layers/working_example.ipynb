# 4.3.2 Pooling Layers

MaxPool, AvgPool, Global Pooling — manual implementations and shape effects.

```python
import numpy as np

def maxpool(x,s=2,st=2):
    oh=(x.shape[0]-s)//st+1; ow=(x.shape[1]-s)//st+1
    return np.array([[x[i*st:i*st+s,j*st:j*st+s].max() for j in range(ow)] for i in range(oh)])

def avgpool(x,s=2,st=2):
    oh=(x.shape[0]-s)//st+1; ow=(x.shape[1]-s)//st+1
    return np.array([[x[i*st:i*st+s,j*st:j*st+s].mean() for j in range(ow)] for i in range(oh)])

fm=np.random.default_rng(42).standard_normal((8,8))
for name,fn in [('MaxPool',maxpool),('AvgPool',avgpool)]:
    out=fn(fm); print(f'{name}: {fm.shape} → {out.shape}')
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,3,figsize=(10,3))
for ax,arr,t in zip(axes,[fm,maxpool(fm),avgpool(fm)],['Input 8x8','MaxPool 4x4','AvgPool 4x4']):
    ax.imshow(arr,cmap='viridis'); ax.set_title(t); ax.axis('off')
plt.tight_layout(); plt.show()
```
