# 4.3.3 CNN Architectures (Historical Evolution)

LeNet → AlexNet → VGG → GoogLeNet → ResNet — key innovations timeline.

```python
# Architecture comparison table
archs = [
    ('LeNet-5 (1998)',   '5×5 Conv×2, AvgPool, Tanh',        'Digit recognition'),
    ('AlexNet (2012)',  '11×11→5×5→3×3, ReLU, Dropout',     'ImageNet winner'),
    ('VGG-16 (2014)',   '3×3 Conv stacks, simple depth',     'Easy to understand'),
    ('GoogLeNet (2014)','Inception: parallel 1×1/3×3/5×3',  'Efficient depth'),
    ('ResNet-50 (2015)','Skip connections, 50 layers',        'Deep without vanishing'),
]
print(f'{"Arch":20s} {"Key Innovation":35s} {"Why Important":25s}')
print('-'*80)
for a,k,w in archs: print(f'{a:20s} {k:35s} {w:25s}')
```
