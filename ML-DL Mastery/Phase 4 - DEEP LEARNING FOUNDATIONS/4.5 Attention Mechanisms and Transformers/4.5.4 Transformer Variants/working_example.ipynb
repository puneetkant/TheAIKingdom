# 4.5.4 Transformer Variants

BERT (encoder-only) vs GPT (decoder-only) vs T5 (encoder-decoder) — when to use what.

```python
# Architecture comparison
for name,mask,use in [
    ('BERT','Bidirectional','Classification, NER, QA'),
    ('GPT', 'Causal',       'Text generation, chat, code'),
    ('T5',  'Enc-bidi/Dec-causal','Translation, summarization'),
]:
    print(f'{name:6s}: mask={mask:25s} → {use}')
```

```python
import numpy as np, matplotlib.pyplot as plt
n=5
fig,axes=plt.subplots(1,2,figsize=(8,3))
axes[0].imshow(np.tril(np.ones((n,n))),cmap='Blues',vmin=0,vmax=1); axes[0].set_title('GPT: Causal Mask')
axes[1].imshow(np.ones((n,n)),cmap='Blues',vmin=0,vmax=1); axes[1].set_title('BERT: Bidirectional')
for ax in axes: ax.set_xlabel('Key'); ax.set_ylabel('Query')
plt.tight_layout(); plt.show()
```
