# 4.5.1 Attention Fundamentals

Scaled dot-product attention: Attention(Q,K,V) = softmax(QKᵀ/√dₖ)·V

```python
import numpy as np

def softmax(x,axis=-1):
    x=x-x.max(axis=axis,keepdims=True); e=np.exp(x); return e/e.sum(axis=axis,keepdims=True)

def attention(Q,K,V,mask=None):
    dk=Q.shape[-1]; s=Q@K.T/np.sqrt(dk)
    if mask is not None: s=np.where(mask,-1e9,s)
    w=softmax(s); return w@V, w

rng=np.random.default_rng(42); L,dk,dv=5,8,8
Q=rng.standard_normal((L,dk)); K=rng.standard_normal((L,dk)); V=rng.standard_normal((L,dv))
ctx,w=attention(Q,K,V)
print('Weights shape:',w.shape,'  Context shape:',ctx.shape)
print('Row sums:',w.sum(axis=1).round(4))
```

```python
import matplotlib.pyplot as plt
fig,axes=plt.subplots(1,2,figsize=(10,4))
im=axes[0].imshow(w,cmap='Blues',vmin=0,vmax=1); axes[0].set_title('Attention Weights'); plt.colorbar(im,ax=axes[0])
axes[1].imshow(ctx,cmap='RdBu',aspect='auto'); axes[1].set_title('Context Vectors')
plt.tight_layout(); plt.show()
```

```python
# Causal mask (autoregressive)
mask=np.triu(np.ones((5,5),dtype=bool),k=1)
_,wc=attention(Q,K,V,mask=mask)
print('Causal weights (lower triangle only):\n',np.round(wc,3))
```
