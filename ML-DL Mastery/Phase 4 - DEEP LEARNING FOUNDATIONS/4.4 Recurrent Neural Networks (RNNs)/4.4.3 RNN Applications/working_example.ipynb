# 4.4.3 RNN Applications

Sequence classification (sin vs cos), time series forecasting, sentiment classification sketch.

```python
import numpy as np
sigmoid=lambda x:1/(1+np.exp(-np.clip(x,-500,500)))

def make_seqs(n=300,L=15):
    rng=np.random.default_rng(0); X,y=[],[]
    for _ in range(n):
        ph,fr=rng.uniform(0,2*np.pi),rng.uniform(.5,2.); t=np.linspace(0,2*np.pi,L)
        c=rng.integers(0,2); s=np.sin(fr*t+ph) if c==0 else np.cos(fr*t+ph)
        X.append(s.reshape(-1,1)+rng.standard_normal(L).reshape(-1,1)*.2); y.append(c)
    return X,np.array(y)

X,y=make_seqs(); print(f'Sequences: {len(X)}, each shape: {X[0].shape}')
```

```python
# RNN inference (no training here for brevity)
rng=np.random.default_rng(42); hid=16
Wxh=rng.standard_normal((1,hid))*.1; Whh=rng.standard_normal((hid,hid))*.1; bh=np.zeros(hid)
Why=rng.standard_normal((hid,1))*.1; by=np.zeros(1)
preds=[]
for xs in X:
    h=np.zeros(hid)
    for x in xs: h=np.tanh(x@Wxh+h@Whh+bh)
    preds.append(int(sigmoid(h@Why+by)[0]>0.5))
acc=sum(p==t for p,t in zip(preds,y))/len(y)
print(f'Random-weight accuracy: {acc:.4f}  (expect ~0.5)')
```
