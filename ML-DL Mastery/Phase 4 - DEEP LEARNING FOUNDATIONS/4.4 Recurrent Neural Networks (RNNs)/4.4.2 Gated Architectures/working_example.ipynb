# 4.4.2 Gated Architectures (LSTM & GRU)

Forget/input/output gates in LSTM; reset/update gates in GRU — from scratch.

```python
import numpy as np
sigmoid=lambda x:1/(1+np.exp(-np.clip(x,-500,500)))

class GRU:
    def __init__(self,inp,hid,seed=42):
        r=np.random.default_rng(seed); s=.1; n=inp+hid
        self.Wz=r.standard_normal((n,hid))*s; self.bz=np.zeros(hid)
        self.Wr=r.standard_normal((n,hid))*s; self.br=np.zeros(hid)
        self.Wn=r.standard_normal((n,hid))*s; self.bn=np.zeros(hid)
        self.h=np.zeros(hid)
    def step(self,x):
        xh=np.concatenate([x,self.h])
        z=sigmoid(xh@self.Wz+self.bz); r=sigmoid(xh@self.Wr+self.br)
        n=np.tanh(np.concatenate([x,r*self.h])@self.Wn+self.bn)
        self.h=(1-z)*self.h+z*n; return self.h.copy()

t=np.linspace(0,8*np.pi,100); gru=GRU(1,8)
hs=np.array([gru.step(np.array([x])) for x in np.sin(t)])
print('GRU hidden state: shape=',hs.shape,'  mean=',hs.mean().round(4))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.imshow(hs.T,aspect='auto',cmap='RdBu',vmin=-1,vmax=1)
plt.colorbar(); plt.title('GRU Hidden States (8 units over 100 time steps)'); plt.xlabel('Time'); plt.show()
```
