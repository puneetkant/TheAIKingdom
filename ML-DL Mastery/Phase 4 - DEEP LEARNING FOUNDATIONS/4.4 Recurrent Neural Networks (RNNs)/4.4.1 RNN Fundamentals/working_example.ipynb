# 4.4.1 RNN Fundamentals

Vanilla RNN: h_t = tanh(W_xh·x_t + W_hh·h_{t-1} + b). Sine wave prediction demo.

```python
import numpy as np

class VanillaRNN:
    def __init__(self,inp,hid,out,seed=42):
        r=np.random.default_rng(seed); s=.01
        self.Wxh=r.standard_normal((inp,hid))*s; self.Whh=r.standard_normal((hid,hid))*s
        self.bh=np.zeros(hid); self.Why=r.standard_normal((hid,out))*s; self.by=np.zeros(out)
        self.h=np.zeros(hid)
    def step(self,x):
        self.h=np.tanh(x@self.Wxh+self.h@self.Whh+self.bh)
        return self.h@self.Why+self.by

rnn=VanillaRNN(1,16,1)
t=np.linspace(0,4*np.pi,100); s=np.sin(t)
for x in s[:5]: print(f'x={x:.3f}  y={rnn.step(np.array([x]))[0]:.4f}')
```

```python
import matplotlib.pyplot as plt
rnn2=VanillaRNN(1,16,1); preds=[]
for x in s: preds.append(rnn2.step(np.array([x]))[0])
plt.plot(s,label='True'); plt.plot(preds,label='RNN (random)'); plt.legend(); plt.title('Vanilla RNN Sine'); plt.show()
```
